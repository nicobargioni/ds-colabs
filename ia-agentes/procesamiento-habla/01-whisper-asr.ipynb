Conseguimos un clip de voz: descarga de un sample público (opción A) o subí el tuyo (opción B).

```python
# Obtener un audio de ejemplo
# OPCIÓN A (por defecto): descargar un sample público corto de habla en inglés.
#   Es un clip de dominio público que viene con el repo de librosa (voz + breve).
# OPCIÓN B (alternativa): grabar tu propia voz / subir un archivo. Descomentá el bloque de abajo.

AUDIO_PATH = 'sample.wav'

# --- OPCIÓN A: descargar sample público (LibriVox / dominio público) ---
# Usamos el ejemplo 'libri1' incluido en librosa: una grabación de voz hablada, corta y limpia.
y, sr = librosa.load(librosa.example('libri1'), sr=16000)   # Whisper espera 16 kHz
import soundfile as sf
sf.write(AUDIO_PATH, y, sr)
dur = len(y) / sr
print(f'Audio listo: {AUDIO_PATH} · {dur:.1f} s · sample rate {sr} Hz · {len(y)} muestras')

# --- OPCIÓN B: subir tu propio audio (descomentar para usar) ---
# from google.colab import files
# up = files.upload()                 # elegí un .wav o .mp3 desde tu compu
# AUDIO_PATH = next(iter(up))
# y, sr = librosa.load(AUDIO_PATH, sr=16000)

# Reproductor para escuchar el clip dentro del notebook
from IPython.display import Audio, display
display(Audio(y, rate=sr))
```

## 🌈 2 · El sonido como imagen

Visualizamos el audio como espectrograma log-Mel: la representación tiempo-frecuencia que consume el encoder.

```python
# El sonido como imagen: espectrograma log-Mel
# Whisper NO ve la onda cruda: la convierte a un espectrograma log-Mel.
# - Eje X = tiempo, Eje Y = frecuencia (banco de filtros Mel ~ percepción humana),
# - color = energía (en dB) de cada frecuencia en cada instante.
S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=80, n_fft=400, hop_length=160)
S_db = librosa.power_to_db(S, ref=np.max)

fig, ax = plt.subplots(2, 1, figsize=(11, 6), sharex=True)
ax[0].plot(np.arange(len(y)) / sr, y, color='#7fd4ff', lw=0.6)
ax[0].set_title('Onda cruda (amplitud vs tiempo)'); ax[0].set_ylabel('amplitud')
img = librosa.display.specshow(S_db, sr=sr, hop_length=160, x_axis='time', y_axis='mel',
                               ax=ax[1], cmap='magma')
ax[1].set_title('Espectrograma log-Mel (lo que "ve" el encoder de Whisper)')
fig.colorbar(img, ax=ax[1], format='%+2.0f dB')
plt.tight_layout(); plt.show()
print(f'Espectrograma: {S_db.shape[0]} bandas Mel × {S_db.shape[1]} frames temporales')
```

## ✍️ 3 · Transcripción

Cargamos `whisper.load_model('base')`, transcribimos y mostramos el texto y los segmentos con timestamps.

```python
# Transcripción con Whisper
# Cargamos el modelo "base" (~74M parámetros): buen equilibrio calidad/velocidad.
# Opciones: tiny < base < small < medium < large (más grande = mejor y más lento).
model = whisper.load_model('base', device=DEVICE)
print('Modelo cargado:', sum(p.numel() for p in model.parameters()) / 1e6, 'M parámetros')

result = model.transcribe(AUDIO_PATH, fp16=(DEVICE == 'cuda'))

print('\n=== TEXTO TRANSCRIPTO ===')
print(result['text'].strip())

print('\n=== SEGMENTOS (con timestamps) ===')
for seg in result['segments']:
    print(f"[{seg['start']:6.2f}s → {seg['end']:6.2f}s]  {seg['text'].strip()}")
```

**¿Qué es ASR y por qué Whisper es robusto?**

El *Automatic Speech Recognition* (ASR) convierte una señal de audio en texto. El enfoque clásico encadenaba módulos separados (modelo acústico + diccionario fonético + modelo de lenguaje), cada uno entrenado por su lado y frágil ante acentos o ruido. **Whisper** reemplaza todo eso por un único **Transformer encoder-decoder** (seq2seq) entrenado de punta a punta sobre **680.000 horas** de audio recolectado de la web, multilingüe y débilmente supervisado. Esa escala y diversidad son la razón de su robustez: vio tantos acentos, dominios y condiciones de grabación que generaliza a audio nuevo **sin fine-tuning**.

**Cómo procesa el audio.** La onda se convierte en un **espectrograma log-Mel** (la imagen que graficamos arriba). El *encoder* lee ese espectrograma y produce una representación; el *decoder* genera el texto **token por token** de forma autorregresiva, condicionado por el encoder.

**Tokens de tarea (multitarea).** Whisper no solo transcribe: con los mismos pesos puede **traducir a inglés**, **detectar idioma** y marcar timestamps. Lo que decide la tarea son **tokens especiales** que se inyectan al inicio de la secuencia del decoder, por ejemplo `<|startoftranscript|>`, `<|es|>` (idioma), `<|transcribe|>` vs `<|translate|>`, `<|notimestamps|>`. Cambiar esos tokens cambia el comportamiento sin reentrenar nada.

**Limitaciones.** En tramos de **silencio o ruido sin habla**, el decoder autorregresivo tiende a **alucinar** texto (repite frases o inventa contenido), porque siempre intenta emitir tokens plausibles. También puede equivocar nombres propios poco frecuentes y degrada en idiomas con poca representación en el corpus. Mitigaciones habituales: usar `condition_on_previous_text=False`, recortar silencios con VAD, o subir el tamaño del modelo (`small`/`medium`) cuando hay GPU disponible.

## 🌐 4 · Detección de idioma (opcional)

Con los mismos pesos, Whisper estima el idioma a partir de los primeros 30 s de audio.

```python
# (Opcional) Detección de idioma
# Whisper detecta el idioma analizando los primeros 30 s de audio antes de transcribir.
audio = whisper.load_audio(AUDIO_PATH)
audio = whisper.pad_or_trim(audio)                 # Whisper trabaja en ventanas de 30 s
mel = whisper.log_mel_spectrogram(audio).to(model.device)

_, probs = model.detect_language(mel)
top = sorted(probs.items(), key=lambda kv: kv[1], reverse=True)[:5]
print('Idioma más probable:', max(probs, key=probs.get))
print('Top-5 candidatos:')
for lang, p in top:
    print(f'  {lang:>4}  {p:.3f}')
```

Si usás tu propio audio, prestá atención a si aparece texto en segmentos donde no hablás: eso sería una **alucinación** en silencio, la limitación descrita arriba.

## ✅ Conclusiones

- Whisper transcribe habla **sin fine-tuning**: el preentrenamiento a gran escala (680k h) le da robustez ante acentos, ruido y dominios variados.
- El audio se procesa como **espectrograma log-Mel** (imagen tiempo-frecuencia): el encoder ve patrones espectrales, no la onda cruda.
- Es **multitarea**: los mismos pesos transcriben, traducen y detectan idioma, controlados por **tokens de tarea** en el decoder.
- Limitación clave: en **silencios o ruido sin habla** el decoder puede **alucinar** texto; conviene validar segmentos vacíos y bajar el tamaño del modelo si no hay GPU.
