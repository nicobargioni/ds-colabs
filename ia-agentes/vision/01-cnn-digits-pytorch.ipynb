## 🧭 Marco del trabajo

| | |
|---|---|
| **Objetivo** | Entrenar una CNN chica en CPU para clasificar dígitos 8×8 y compararla contra un baseline de regresión logística sobre los píxeles planos. |
| **Hipótesis** | La CNN, al explotar la estructura espacial 2D con filtros convolucionales, iguala o supera a un clasificador lineal que trata cada píxel como una feature independiente. |
| **Datos** | Las 1797 imágenes 8×8 (16 niveles de gris) del dataset load_digits de scikit-learn, una versión reducida del clásico MNIST de la UCI. |
| **Metodología** | load_digits → split estratificado (random_state=42) → normalización → tensores 1×8×8 → CNN entrenada con Adam/CrossEntropy → evaluación (accuracy, matriz de confusión) vs baseline LogisticRegression. |
| **Métrica** | Accuracy en test + matriz de confusión |

> Ejecutable top-down · `random_state=42` · preprocesamiento **sin data leakage**.

## 🛠 0 · Setup

Imports (PyTorch, scikit-learn, Plotly), semilla fija y device CPU.

```python
import warnings; warnings.filterwarnings('ignore')
import numpy as np
import torch, torch.nn as nn, torch.nn.functional as F
from torch.utils.data import TensorDataset, DataLoader
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
import plotly.express as px, plotly.graph_objects as go, plotly.io as pio
from plotly.subplots import make_subplots
pio.templates.default = 'plotly_dark'

SEED = 42
torch.manual_seed(SEED); np.random.seed(SEED)
DEVICE = torch.device('cpu')          # corre 100% en CPU; no requiere GPU
PRIMARY, ACCENT = '#7fd4ff', '#ff9f5c'
print('Setup OK · PyTorch', torch.__version__, '· device =', DEVICE)
print('Hilos CPU disponibles para torch:', torch.get_num_threads())
```

## 🔢 1 · Datos

Cargamos `load_digits` (1797 imágenes 8×8), las miramos, hacemos el split estratificado, normalizamos sin leakage y armamos tensores 1×8×8.

```python
# 1a · Cargar el dataset y mirar su forma
digits = load_digits()
X_img, y = digits.images, digits.target          # X_img: (1797, 8, 8) en [0, 16]
print('Imágenes:', X_img.shape, '· rango de intensidad:', X_img.min(), '-', X_img.max())
print('Etiquetas:', y.shape, '· clases:', np.unique(y))
print('Ejemplos por clase:', np.bincount(y))      # dataset balanceado
```

```python
# 1b · Mostrar un ejemplo de cada dígito (heatmaps Plotly)
fig = make_subplots(rows=2, cols=5,
                    subplot_titles=[f'dígito {d}' for d in range(10)],
                    horizontal_spacing=0.04, vertical_spacing=0.12)
for d in range(10):
    idx = np.where(y == d)[0][0]                  # primer ejemplo de cada clase
    r, c = divmod(d, 5)
    fig.add_trace(go.Heatmap(z=X_img[idx][::-1], colorscale='Blues',
                             showscale=False), row=r+1, col=c+1)
fig.update_xaxes(visible=False); fig.update_yaxes(visible=False)
fig.update_layout(height=420, title='Un ejemplo de cada dígito (8×8 píxeles, 0–16 niveles de gris)')
fig.show()
```

```python
# 1c · Split estratificado + normalización + tensores 1×8×8
# Split ANTES de calcular estadísticas → sin data leakage.
X_flat = X_img.reshape(len(X_img), -1)            # (1797, 64) para el baseline y el split
Xtr_f, Xte_f, ytr, yte = train_test_split(
    X_flat, y, test_size=0.25, stratify=y, random_state=SEED)

# Normalización: media/desvío calculados SOLO con train, aplicados a ambos.
mu, sigma = Xtr_f.mean(), Xtr_f.std()
Xtr_n = (Xtr_f - mu) / sigma
Xte_n = (Xte_f - mu) / sigma

# A tensores con forma (N, C=1, H=8, W=8) que es lo que espera Conv2d.
Xtr_t = torch.tensor(Xtr_n, dtype=torch.float32).reshape(-1, 1, 8, 8)
Xte_t = torch.tensor(Xte_n, dtype=torch.float32).reshape(-1, 1, 8, 8)
ytr_t = torch.tensor(ytr, dtype=torch.long)
yte_t = torch.tensor(yte, dtype=torch.long)

train_dl = DataLoader(TensorDataset(Xtr_t, ytr_t), batch_size=64, shuffle=True)
print(f'Train: {Xtr_t.shape}  ·  Test: {Xte_t.shape}')
print(f'Normalización (solo train) → media={mu:.3f}, desvío={sigma:.3f}')
```

## 🧠 2 · Modelo

Una CNN chica: Conv2d→ReLU→MaxPool → Conv2d→ReLU→MaxPool → Flatten → Linear.

```python
# 2 · La CNN: dos bloques convolucionales + clasificador lineal
class DigitCNN(nn.Module):
    def __init__(self, n_classes=10):
        super().__init__()
        # Conv1: 1 canal de entrada → 8 filtros 3×3, padding=1 conserva el tamaño (8×8).
        #        Cada filtro aprende un detector local (bordes, esquinas, trazos).
        self.conv1 = nn.Conv2d(in_channels=1,  out_channels=8,  kernel_size=3, padding=1)
        # Conv2: 8 → 16 filtros 3×3. Opera sobre el mapa ya reducido (4×4) y combina
        #        las features de bajo nivel en patrones más complejos.
        self.conv2 = nn.Conv2d(in_channels=8,  out_channels=16, kernel_size=3, padding=1)
        self.pool  = nn.MaxPool2d(2)              # reduce a la mitad H y W (downsampling)
        # Tras conv1→pool: 8×(4×4). Tras conv2→pool: 16×(2×2) = 64 valores aplanados.
        self.fc    = nn.Linear(16 * 2 * 2, n_classes)

    def forward(self, x):                         # x: (N, 1, 8, 8)
        x = self.pool(F.relu(self.conv1(x)))      # → (N, 8, 4, 4)
        x = self.pool(F.relu(self.conv2(x)))      # → (N, 16, 2, 2)
        x = x.flatten(1)                          # → (N, 64)
        return self.fc(x)                         # logits → (N, 10)

model = DigitCNN().to(DEVICE)
n_params = sum(p.numel() for p in model.parameters())
print(model)
print(f'\nParámetros entrenables: {n_params:,}')
```

**Anatomía de la red, capa por capa.**

- **`Conv2d(1→8, 3×3)`** desliza 8 filtros de 3×3 sobre la imagen. Cada filtro es un detector local de patrones de bajo nivel (un borde inclinado, una esquina, un trazo). El `padding=1` mantiene el tamaño 8×8 para no perder los bordes. La salida son 8 *mapas de activación* que dicen *dónde* aparece cada patrón.
- **`ReLU`** introduce no linealidad (pone en cero las activaciones negativas). Sin ella, apilar convoluciones colapsaría en una sola transformación lineal.
- **`MaxPool2d(2)`** se queda con el máximo de cada ventana 2×2: reduce la resolución a la mitad (8×8 → 4×4), da algo de invariancia a pequeños desplazamientos y baja el cómputo.
- **`Conv2d(8→16, 3×3)`** actúa sobre los mapas ya reducidos: combina las features simples en patrones más complejos (combinaciones de trazos que forman parte de un dígito). Tras su pooling quedan 16 mapas de 2×2.
- **`Flatten`** estira esos 16×2×2 = 64 valores en un vector.
- **`Linear(64→10)`** es el clasificador: produce un *logit* por clase. El más alto es la predicción.

La idea clave es la **jerarquía de features**: las primeras capas ven detalles locales, las siguientes los combinan en estructuras cada vez más abstractas. Es la diferencia con un modelo lineal sobre píxeles, que trata cada píxel como una variable suelta sin noción de vecindad espacial.

## 🏋️ 3 · Entrenamiento

Loop de 20 épocas en CPU con Adam y CrossEntropy; guardamos loss y accuracy por época.

```python
# 3 · Entrenamiento en CPU (Adam + CrossEntropy)
EPOCHS = 20
opt = torch.optim.Adam(model.parameters(), lr=1e-3)
crit = nn.CrossEntropyLoss()                       # combina log-softmax + NLL

loss_hist, acc_hist = [], []
for ep in range(1, EPOCHS + 1):
    model.train(); running = 0.0
    for xb, yb in train_dl:
        opt.zero_grad()
        loss = crit(model(xb), yb)
        loss.backward(); opt.step()
        running += loss.item() * len(xb)
    epoch_loss = running / len(train_dl.dataset)
    # accuracy en train (rápido, sin gradientes) para ver el aprendizaje
    model.eval()
    with torch.no_grad():
        tr_acc = (model(Xtr_t).argmax(1) == ytr_t).float().mean().item()
    loss_hist.append(epoch_loss); acc_hist.append(tr_acc)
    if ep % 4 == 0 or ep == 1:
        print(f'época {ep:2d}/{EPOCHS} · loss={epoch_loss:.4f} · acc_train={tr_acc:.4f}')
print('\nEntrenamiento terminado.')
```

```python
# 3b · Curva de loss y accuracy de entrenamiento por época
fig = make_subplots(specs=[[{'secondary_y': True}]])
ep_axis = list(range(1, EPOCHS + 1))
fig.add_trace(go.Scatter(x=ep_axis, y=loss_hist, name='Loss (train)',
                         line=dict(color=ACCENT, width=3)), secondary_y=False)
fig.add_trace(go.Scatter(x=ep_axis, y=acc_hist, name='Accuracy (train)',
                         line=dict(color=PRIMARY, width=3)), secondary_y=True)
fig.update_xaxes(title_text='Época')
fig.update_yaxes(title_text='CrossEntropy loss', secondary_y=False)
fig.update_yaxes(title_text='Accuracy', secondary_y=True, range=[0, 1.01])
fig.update_layout(title='Curva de entrenamiento de la CNN (CPU, 20 épocas)', height=420)
fig.show()
```

## 📊 4 · Evaluación

Accuracy en test, matriz de confusión y comparación contra un baseline de regresión logística sobre los píxeles planos.

```python
# 4 · Evaluación en test: CNN vs baseline lineal
# --- CNN ---
model.eval()
with torch.no_grad():
    cnn_pred = model(Xte_t).argmax(1).numpy()
acc_cnn = accuracy_score(yte, cnn_pred)

# --- Baseline: regresión logística sobre los 64 píxeles planos (mismos datos normalizados) ---
logreg = LogisticRegression(max_iter=2000, random_state=SEED)
logreg.fit(Xtr_n, ytr)
base_pred = logreg.predict(Xte_n)
acc_base = accuracy_score(yte, base_pred)

print(f'Accuracy CNN (PyTorch)            : {acc_cnn:.4f}')
print(f'Accuracy baseline (LogisticReg)   : {acc_base:.4f}')
print(f'Diferencia (CNN - baseline)       : {acc_cnn - acc_base:+.4f}')
```

```python
# 4b · Matriz de confusión de la CNN (Plotly heatmap)
cm = confusion_matrix(yte, cnn_pred)
fig = px.imshow(cm, text_auto=True, color_continuous_scale='Blues',
                labels=dict(x='Predicho', y='Real', color='Conteo'),
                x=[str(d) for d in range(10)], y=[str(d) for d in range(10)],
                title=f'Matriz de confusión · CNN (accuracy test = {acc_cnn:.3f})')
fig.update_layout(height=520, coloraxis_showscale=True)
fig.show()

# Resumir los errores fuera de la diagonal (qué se confunde con qué)
errs = [(yte_lbl, pred_lbl, int(cm[yte_lbl, pred_lbl]))
        for yte_lbl in range(10) for pred_lbl in range(10)
        if yte_lbl != pred_lbl and cm[yte_lbl, pred_lbl] > 0]
errs.sort(key=lambda t: -t[2])
print('Confusiones más frecuentes (real → predicho : conteo):')
for real, pred, n in errs[:6]:
    print(f'  {real} → {pred} : {n}')
print(f'Total de errores de la CNN en test: {(cnn_pred != yte).sum()} de {len(yte)}')
```

**Cómo leer la comparación.** La CNN y el baseline se evalúan sobre exactamente el mismo test set y los mismos datos normalizados, así que la diferencia mide solo el aporte de la arquitectura convolucional. La **matriz de confusión** pone los aciertos en la diagonal; todo lo que cae fuera son errores, y su posición dice *qué* dígito se confundió con cuál. Los errores típicos no son aleatorios: aparecen entre dígitos que comparten trazos a resolución 8×8 (donde la imagen es tan chica que algunas formas quedan casi idénticas). Por eso conviene mirar las celdas fuera de la diagonal más que la accuracy global.

### 📌 Lectura de resultados

Entrenamos la CNN **20 épocas en CPU** (1.898 parámetros, ~8 s de ejecución total del notebook) sobre 1.347 imágenes de train y la evaluamos sobre 450 de test. Estos son los números reales:

| Modelo | Accuracy en test | Errores (de 450) |
|---|---|---|
| Baseline · `LogisticRegression` (64 píxeles planos) | **0.9667** | 15 |
| CNN · PyTorch (Conv→ReLU→Pool ×2 → Linear) | **0.9467** | 24 |

**Qué dicen los números (y por qué contradicen la intuición "CNN gana siempre").**

- La curva de entrenamiento es sana: la loss baja de **2.26 → 0.12** y la accuracy de train sube de 0.35 a **0.970** sin estancarse ni dispararse. La CNN *aprende*, no hay un bug.
- Pero en test el **baseline lineal gana por 2 puntos** (96.7% vs 94.7%). La hipótesis —que la convolución igualaría o superaría al modelo lineal— **no se cumple acá**, y el motivo es instructivo: a **8×8 píxeles** el dataset `load_digits` es casi *linealmente separable*. Con solo 64 features y 10 clases bien balanceadas, una regresión logística tiene margen de sobra para trazar fronteras limpias, mientras que la CNN —con su pooling agresivo que reduce 8×8 a 2×2— **tira mucha resolución espacial** justo cuando hay poquísima para empezar. El sesgo inductivo convolucional (localidad, invariancia a traslación) recién paga cuando las imágenes son grandes y los patrones se repiten por toda la escena: MNIST 28×28, y sobre todo imágenes naturales.
- La **matriz de confusión** confirma que los 24 errores no son ruido aleatorio: se concentran en pares de dígitos que comparten trazos a esta resolución diminuta. El error más frecuente es **8 → 1 (4 veces)**, seguido de confusiones recíprocas **1↔2** y **3→2** (2 cada una) y **5→9**. Son justamente los dígitos cuyas formas, comprimidas a 8×8, dejan de ser distinguibles.

**Lectura honesta:** el experimento no "falla" — enseña *cuándo* una CNN aporta y cuándo no. Con datos de juguete y baja resolución, un baseline lineal fuerte es difícil de batir y debe ser siempre el punto de comparación. La ventaja de la CNN se haría visible escalando a 28×28 con torchvision, agregando dropout/batch-norm y data augmentation: ahí la jerarquía de features convolucional empieza a despegar.

## ✅ Conclusiones

- Una **CNN chica** (dos convoluciones + una capa lineal) clasifica dígitos 8×8 con alta precisión entrenando enteramente en **CPU** en segundos.
- Con imágenes tan pequeñas (8×8) el **baseline lineal gana** (96.7% vs 94.7% de la CNN): el dataset es casi linealmente separable, así que la convolución no despega como sí lo haría en MNIST 28×28 o en imágenes naturales.
- La **matriz de confusión** revela los pares de dígitos que comparten trazos (típicamente confusiones entre formas parecidas), no errores aleatorios.
