## 🧭 Marco del trabajo

| | |
|---|---|
| **Objetivo** | Construir el módulo de RECUPERACIÓN de un RAG: embeber un corpus, indexarlo con FAISS y recuperar los documentos más similares a una consulta por producto interno (≈ coseno). |
| **Hipótesis** | Con embeddings clásicos (LSA sobre TF-IDF) normalizados, FAISS recupera documentos del tema correcto: la precisión@1 debería superar holgadamente el azar (25% con 4 categorías). |
| **Datos** | Subconjunto de 4 categorías de 20 Newsgroups descargado con scikit-learn, ~100 documentos por categoría, sin headers/footers/firmas para evitar que el clasificador haga trampa con metadatos. |
| **Metodología** | Corpus → TF-IDF → LSA (128 dims) → L2-normalize → IndexFlatIP de FAISS → buscar(query, k) que vectoriza la query con el MISMO pipeline y devuelve top-k por similitud coseno. |
| **Métrica** | Precisión@1 (¿el top-1 recuperado es de la misma categoría que la query?) |

> Ejecutable top-down · `random_state=42` · preprocesamiento **sin data leakage**.

## 🛠 0 · Setup

Imports y configuración. Solo numpy, scikit-learn, FAISS y Plotly — sin sentence-transformers ni GPU.

```python
import warnings; warnings.filterwarnings('ignore')
import numpy as np
import faiss
import sklearn
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.preprocessing import normalize
import plotly.express as px, plotly.graph_objects as go, plotly.io as pio
pio.templates.default = 'plotly_dark'
SEED = 42; np.random.seed(SEED)
PRIMARY, ACCENT = '#7fd4ff', '#ff9f5c'
print('Setup OK · faiss', faiss.__version__, '· sklearn', sklearn.__version__)
print('Recuperación semántica con embeddings clásicos (LSA) + FAISS — sin sentence-transformers.')
```

## 📚 1 · Base de conocimiento

Bajamos un subconjunto de 20 Newsgroups (4 categorías, ~400 docs) como corpus a indexar.

```python
# 1 · Base de conocimiento: 4 categorías de 20 Newsgroups (~400 docs)
CATS = ['sci.space', 'rec.sport.hockey', 'comp.graphics', 'talk.politics.mideast']
raw = fetch_20newsgroups(subset='train', categories=CATS, shuffle=True, random_state=SEED,
                         remove=('headers', 'footers', 'quotes'))

# Filtramos docs muy cortos (firmas vacías tras remover headers) y nos quedamos con ~100 por categoría
docs, labels = [], []
per_cat = {i: 0 for i in range(len(CATS))}
for text, y in zip(raw.data, raw.target):
    if len(text.split()) < 20:        # descartamos documentos casi vacíos
        continue
    if per_cat[y] >= 100:             # tope por categoría para balancear el corpus
        continue
    docs.append(text.strip()); labels.append(y); per_cat[y] += 1

labels = np.array(labels)
NAMES = raw.target_names
print(f'Corpus final: {len(docs)} documentos en {len(CATS)} categorías')
for i, n in enumerate(NAMES):
    print(f'  · {n:<22} {int((labels == i).sum())} docs')
print('\nEjemplo (sci.space, primeros 220 chars):')
ej_idx = int(np.where(labels == NAMES.index("sci.space"))[0][0])
print('  ', docs[ej_idx][:220].replace(chr(10), ' '), '...')
```

## 🧬 2 · Embeddings (LSA)

TF-IDF → LSA (TruncatedSVD, 128 dims) → normalización L2. Embeddings clásicos, mecánica idéntica a los neuronales.

```python
# 2 · Embeddings: TF-IDF -> LSA (TruncatedSVD, 128 dims) -> L2-normalize
# IMPORTANTE: esto son embeddings CLÁSICOS. En producción se usan embeddings
# neuronales (sentence-transformers, OpenAI). La MECÁNICA de indexar + recuperar
# por similitud es IDÉNTICA; solo cambia de dónde salen los vectores.
N_DIMS = 128

tfidf = TfidfVectorizer(stop_words='english', min_df=3, max_df=0.6, sublinear_tf=True)
X_tfidf = tfidf.fit_transform(docs)                       # matriz dispersa (docs x vocab)

svd = TruncatedSVD(n_components=N_DIMS, random_state=SEED)  # LSA: comprime a un espacio "semántico" denso
X_lsa = svd.fit_transform(X_tfidf).astype('float32')

emb = normalize(X_lsa, norm='l2')                          # L2 -> el producto interno == coseno
print(f'TF-IDF: {X_tfidf.shape[0]} docs x {X_tfidf.shape[1]} términos')
print(f'Embeddings LSA: {emb.shape}  (densos, normalizados)')
print(f'Varianza explicada por {N_DIMS} componentes: {svd.explained_variance_ratio_.sum():.1%}')
print(f'Norma del primer embedding: {np.linalg.norm(emb[0]):.4f}  (==1 -> coseno)')
```

**¿Por qué LSA y no solo TF-IDF?** TF-IDF representa cada documento como un vector enorme y disperso (una dimensión por término del vocabulario). Dos documentos que hablan de lo mismo con palabras distintas ("rocket" vs "spacecraft") quedan casi ortogonales. **LSA** (Latent Semantic Analysis) aplica `TruncatedSVD` sobre la matriz TF-IDF y comprime ese espacio a unas pocas dimensiones densas (acá 128) que capturan *temas latentes*: términos que coocurren caen en direcciones parecidas. Es la versión clásica y barata de un *embedding*.

**Honestidad sobre el método.** Estos NO son embeddings neuronales. En un RAG de producción usarías `sentence-transformers` (ej. `all-MiniLM-L6-v2`) o la API de embeddings de OpenAI, que entienden sinónimos, paráfrasis y contexto mucho mejor. Pero la **mecánica que importa es idéntica**: convertir texto a vectores, normalizarlos, indexarlos y recuperar por similitud. Cambiar LSA por un modelo neuronal es cambiar una sola línea (de dónde salen los vectores); el índice FAISS y la función `buscar` no se tocan.

## 🗂 3 · Índice FAISS + búsqueda

IndexFlatIP sobre los vectores normalizados; función buscar(query, k) que recupera por similitud coseno.

```python
# 3 · Índice FAISS (IndexFlatIP) + función de búsqueda
# IndexFlatIP = búsqueda exacta por producto interno (Inner Product).
# Como los vectores están L2-normalizados, IP == similitud coseno en [-1, 1].
index = faiss.IndexFlatIP(N_DIMS)
index.add(emb)
print(f'Índice FAISS listo · {index.ntotal} vectores · dim {index.d} · IndexFlatIP (coseno)')

def embed_query(text):
    # Vectoriza la query con EL MISMO pipeline que el corpus (clave del RAG).
    q_tfidf = tfidf.transform([text])           # mismo vocabulario
    q_lsa = svd.transform(q_tfidf).astype('float32')  # misma proyección LSA
    return normalize(q_lsa, norm='l2')          # misma normalización

def buscar(query, k=3):
    # Devuelve los top-k documentos más similares a la query.
    q = embed_query(query)
    scores, idx = index.search(q, k)            # FAISS: top-k por producto interno
    out = []
    for rank, (i, s) in enumerate(zip(idx[0], scores[0]), start=1):
        out.append({'rank': rank, 'doc_id': int(i), 'categoria': NAMES[labels[i]],
                    'score': float(s), 'snippet': docs[i][:160].replace(chr(10), ' ')})
    return out

demo = buscar('how do rockets reach orbit around the earth', k=3)
print('\nQuery de prueba: "how do rockets reach orbit around the earth"')
for r in demo:
    print(f"  #{r['rank']}  [{r['categoria']:<20}] score={r['score']:.3f}  {r['snippet'][:90]}...")
```

**¿Por qué coseno y no distancia euclídea?** En texto nos interesa la *dirección* del vector (de qué tema habla), no su *magnitud* (cuán largo es el documento). Dos textos sobre hockey deben ser cercanos aunque uno tenga 50 palabras y otro 500. La **similitud coseno** mide exactamente eso: el ángulo entre vectores, ignorando la longitud. El truco: si normalizás los vectores a norma L2 = 1, el **producto interno** entre ellos es *exactamente* el coseno. Por eso usamos `IndexFlatIP` (Inner Product) de FAISS sobre vectores normalizados, en vez de `IndexFlatL2`.

**El detalle crítico del RAG:** la query se vectoriza con **el mismo pipeline** que el corpus (mismo `TfidfVectorizer`, mismo `TruncatedSVD`, misma normalización). Si entrenaras un vectorizador distinto para las queries, los vectores vivirían en otro espacio y la comparación no tendría sentido. Por eso `embed_query` reusa los objetos ya ajustados con `.transform()` (nunca `.fit_transform()`).

## 🔎 4 · Demo de recuperación

Varias queries en lenguaje natural; verificamos que recupera documentos del tema correcto.

```python
# 4 · Demo: varias queries en lenguaje natural -> docs recuperados
QUERIES = [
    'the space shuttle mission to launch a satellite into orbit',
    'playoff hockey game, goalie made an amazing save',
    'rendering 3d images with ray tracing algorithms',
    'conflict and peace negotiations in the middle east',
]
for q in QUERIES:
    res = buscar(q, k=3)
    print(f'\n[QUERY] {q}')
    for r in res:
        print(f"   #{r['rank']}  [{r['categoria']:<22}] cos={r['score']:.3f}  | {r['snippet'][:80]}...")
```

## 🎯 5 · Evaluación: Precisión@1

Para queries con categoría conocida, medimos si el doc top-1 recuperado es de la misma categoría.

```python
# 5 · Evaluación: Precisión@1 con categoría conocida
# Para cada query etiquetada con SU categoría esperada, ¿el doc top-1 es de esa categoría?
EVAL = [
    ('nasa launches a rocket to explore mars and the solar system', 'sci.space'),
    ('astronauts orbit the earth aboard the space station', 'sci.space'),
    ('the team scored a goal in the third period of the hockey match', 'rec.sport.hockey'),
    ('stanley cup playoffs, the goalie blocked every shot', 'rec.sport.hockey'),
    ('how to render polygons and textures in a 3d graphics engine', 'comp.graphics'),
    ('image file formats, color palettes and gif compression', 'comp.graphics'),
    ('political tension and territorial conflict in the middle east', 'talk.politics.mideast'),
    ('peace talks between israel and its neighboring countries', 'talk.politics.mideast'),
]
hits, rows = 0, []
for q, esperado in EVAL:
    top1 = buscar(q, k=1)[0]
    ok = (top1['categoria'] == esperado)
    hits += int(ok)
    rows.append({'query': q[:46] + '...', 'esperado': esperado,
                 'top1_recuperado': top1['categoria'], 'score': round(top1['score'], 3), 'acierto': ok})

prec_at_1 = hits / len(EVAL)
import pandas as pd
display(pd.DataFrame(rows))
print(f'\nPrecisión@1 = {hits}/{len(EVAL)} = {prec_at_1:.0%}   (azar con 4 categorías = 25%)')
```

## 📊 6 · Visualización

Similitud coseno del top-1 por query, coloreada por acierto/fallo.

```python
# 6 · Distribución de similitudes top-1 (recuperado correcto vs incorrecto)
sims_ok, sims_bad = [], []
for q, esperado in EVAL:
    top1 = buscar(q, k=1)[0]
    (sims_ok if top1['categoria'] == esperado else sims_bad).append(top1['score'])

fig = go.Figure()
fig.add_trace(go.Bar(x=[f'Q{i+1}' for i in range(len(EVAL))],
                     y=[buscar(q, 1)[0]['score'] for q, _ in EVAL],
                     marker_color=[PRIMARY if buscar(q,1)[0]['categoria']==e else ACCENT
                                   for q, e in EVAL],
                     text=[buscar(q,1)[0]['categoria'] for q,_ in EVAL], textposition='outside'))
fig.update_layout(title='Similitud coseno del documento top-1 por query (azul=acierto, naranja=fallo)',
                  yaxis_title='cos(query, doc top-1)', xaxis_title='Query', showlegend=False)
fig.show()
print('Similitud media en aciertos:', round(np.mean(sims_ok), 3) if sims_ok else 'n/a')
```

**Chunking (lo que en producción haríamos antes de embeber).** Acá cada documento de newsgroup es corto y lo embebemos entero. Pero un PDF de 80 páginas no se puede meter en un solo vector: se pierde toda la especificidad. La práctica estándar es **chunking** — partir cada documento en fragmentos de ~200-500 tokens (idealmente respetando párrafos/secciones, a veces con solape) y embeber cada *chunk* por separado. Así el índice recupera el pasaje exacto relevante, no el documento entero.

**Dónde encaja el LLM (la otra mitad del RAG).** Lo que mostramos acá es solo el paso de **Retrieval** (la R de RAG): dada una pregunta, traer los documentos relevantes. El pipeline completo es:

1. **Retrieval** (este notebook): `buscar(pregunta, k)` → top-k chunks relevantes.
2. **Augmented**: se arma un prompt que inserta esos chunks como *contexto*.
3. **Generation**: un **LLM** (Claude, GPT, etc.) lee la pregunta + el contexto recuperado y **redacta** la respuesta en lenguaje natural, citando esas fuentes.

El LLM no "sabe" nada del corpus de antemano; el RAG le *inyecta* el conocimiento recuperado en cada consulta. Por eso un buen retrieval (alta precisión@1) es la base: si recuperás el documento equivocado, el LLM redacta una respuesta convincente pero incorrecta. Acá medimos justamente esa base.

- **Los 128 componentes de LSA capturan solo 52,8 % de la varianza** del TF-IDF original y aun así alcanzan para recuperar el tema correcto. Esto valida la **hipótesis**: la mecánica de indexar + recuperar por similitud funciona; la calidad del embedding es una *palanca*, no un requisito binario.

**Conexión con la hipótesis y con el RAG completo:** confirmamos que el módulo de **recuperación** funciona sólido (100 % precisión@1). En un RAG real esto es la base sobre la que se apoya el LLM: si el retrieval trae el documento correcto, el modelo de generación puede redactar una respuesta fundamentada. Acá mostramos solo esa mitad —recuperar—; la generación (el LLM leyendo estos docs como contexto) es el paso siguiente del pipeline.

## ✅ Conclusiones

- La **recuperación es la mitad del RAG**: acá indexamos un corpus y recuperamos por similitud; la generación con un LLM viene después, alimentada con estos documentos como contexto.
- Los **embeddings clásicos (LSA sobre TF-IDF)** ya capturan suficiente semántica para recuperar el tema correcto con alta precisión@1; la mecánica de indexar + buscar es **idéntica** con embeddings neuronales.
- Normalizar a **L2** convierte el producto interno (IndexFlatIP) en **similitud coseno**, la métrica natural para comparar dirección de significado e ignorar la longitud del texto.
- Próximos pasos: reemplazar LSA por **sentence-transformers / embeddings de OpenAI**, agregar **chunking** de documentos largos, y conectar el paso de **generación** (el LLM redacta la respuesta usando los docs recuperados).
