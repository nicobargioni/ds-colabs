## 🧰 1 · Definir las herramientas

Tres tools reales en Python: una calculadora segura (sin `eval`), una búsqueda en un mini-corpus local y la fecha/hora actual.

```python
# ───────────────────────────────────────────────────────────────────────────
# 3 herramientas REALES en Python. Cada una es una función pura y determinista.
# El modelo NO ejecuta este código: nos pide ejecutarlo vía function-calling.
# ───────────────────────────────────────────────────────────────────────────

# --- Tool 1: calculadora SEGURA -------------------------------------------------
# NO usamos eval() (ejecutaría código arbitrario). Evaluamos el AST de la
# expresión permitiendo solo operadores aritméticos.
_OPS = {
    ast.Add: operator.add, ast.Sub: operator.sub, ast.Mult: operator.mul,
    ast.Div: operator.truediv, ast.Pow: operator.pow, ast.Mod: operator.mod,
    ast.USub: operator.neg, ast.UAdd: operator.pos,
}
def _eval_node(node):
    if isinstance(node, ast.Constant) and isinstance(node.value, (int, float)):
        return node.value
    if isinstance(node, ast.BinOp) and type(node.op) in _OPS:
        return _OPS[type(node.op)](_eval_node(node.left), _eval_node(node.right))
    if isinstance(node, ast.UnaryOp) and type(node.op) in _OPS:
        return _OPS[type(node.op)](_eval_node(node.operand))
    raise ValueError("expresión no permitida")

def calculadora(expresion: str) -> str:
    """Evalúa una expresión aritmética (+ - * / ** %) de forma segura."""
    try:
        tree = ast.parse(expresion, mode="eval")
        return str(_eval_node(tree.body))
    except Exception as e:
        return f"ERROR: {e}"

# --- Tool 2: búsqueda en un mini-corpus local ----------------------------------
CORPUS = {
    "nomadic": "Nomadic es una agencia de SEO y producto con foco en automatización e IA.",
    "react": "ReAct (Reason+Act) intercala pasos de razonamiento con llamadas a herramientas.",
    "function calling": "Function-calling: el LLM devuelve una llamada estructurada (nombre + args JSON) que ejecuta tu código.",
    "colab": "Google Colab ejecuta notebooks en la nube con GPU/TPU gratis y manejo de secrets.",
}
def buscar(consulta: str) -> str:
    """Busca en un mini-corpus local por coincidencia de palabras clave."""
    q = consulta.lower()
    hits = [v for k, v in CORPUS.items() if k in q or any(w in k for w in q.split())]
    return " ".join(hits) if hits else "Sin resultados en el corpus."

# --- Tool 3: fecha/hora actual --------------------------------------------------
_DIAS = ["lunes", "martes", "miércoles", "jueves", "viernes", "sábado", "domingo"]
def fecha_hora() -> str:
    """Devuelve la fecha y hora actuales, incluyendo el día de la semana."""
    now = datetime.datetime.now()
    return f"{now:%Y-%m-%d %H:%M} ({_DIAS[now.weekday()]})"

# --- Registro: nombre -> función Python ---------------------------------------
TOOL_FUNCS = {"calculadora": calculadora, "buscar": buscar, "fecha_hora": fecha_hora}

# Prueba rápida (esto SÍ lo corremos nosotros, no el modelo)
print("calculadora('17/100*1480') =", calculadora("17/100*1480"))
print("fecha_hora()              =", fecha_hora())
print("buscar('que es react')    =", buscar("que es react"))
```

## 📐 2 · JSON schema (el contrato)

Cada tool necesita un schema que le diga al modelo qué hace y qué argumentos espera. Es lo que habilita el function-calling.

```python
# ───────────────────────────────────────────────────────────────────────────
# JSON schema de cada tool: es el CONTRATO que el modelo lee para saber
# qué herramientas existen, qué hacen y qué argumentos esperan.
# ───────────────────────────────────────────────────────────────────────────
TOOLS = [
    {
        "name": "calculadora",
        "description": "Evalúa una expresión aritmética (suma, resta, multiplicación, "
                       "división, potencia, módulo). Usala para CUALQUIER cálculo numérico.",
        "input_schema": {
            "type": "object",
            "properties": {
                "expresion": {"type": "string",
                              "description": "Expresión, ej: '17/100*1480' o '(3+4)**2'"},
            },
            "required": ["expresion"],
        },
    },
    {
        "name": "buscar",
        "description": "Busca información en un mini-corpus local de definiciones. "
                       "Usala cuando te pregunten qué es algo del dominio (nomadic, react, colab...).",
        "input_schema": {
            "type": "object",
            "properties": {
                "consulta": {"type": "string", "description": "Términos a buscar"},
            },
            "required": ["consulta"],
        },
    },
    {
        "name": "fecha_hora",
        "description": "Devuelve la fecha y hora actuales y el día de la semana. "
                       "Usala cuando te pregunten por el día de hoy, la hora o la fecha.",
        "input_schema": {"type": "object", "properties": {}},
    },
]
print(f"{len(TOOLS)} herramientas declaradas:", [t["name"] for t in TOOLS])
```

El **schema** es el único lugar donde el modelo aprende qué herramientas tiene. La `description` importa tanto como la firma: el modelo decide *cuándo* usar cada tool leyendo esa descripción, así que conviene ser explícito sobre el cuándo, no solo el qué.

## 🔁 3 · El loop ReAct

Implementamos el ciclo razonar→actuar→observar a mano, con tope de iteraciones para evitar loops infinitos.

```python
# ───────────────────────────────────────────────────────────────────────────
# EL LOOP ReAct (razonar → actuar → observar), implementado a mano.
# Mantenemos control total: límite de iteraciones, logging de la traza.
# ───────────────────────────────────────────────────────────────────────────
def agente_react(pregunta: str, max_iteraciones: int = 6, verbose: bool = True) -> str:
    messages = [{"role": "user", "content": pregunta}]

    for i in range(max_iteraciones):
        if verbose:
            print(f"\n── Iteración {i+1}/{max_iteraciones} ──────────────")

        # 1) RAZONAR: el modelo decide si responde o pide una tool
        resp = client.messages.create(
            model=MODEL, max_tokens=1024, tools=TOOLS, messages=messages,
        )

        # 2) ¿Terminó? Si no pidió tools, devolvemos la respuesta final.
        if resp.stop_reason != "tool_use":
            final = next((b.text for b in resp.content if b.type == "text"), "")
            if verbose:
                print("✔ Respuesta final del modelo")
            return final

        # Guardamos el turno del asistente (incluye los bloques tool_use)
        messages.append({"role": "assistant", "content": resp.content})

        # 3) ACTUAR + OBSERVAR: ejecutamos cada tool que pidió y devolvemos el resultado
        tool_results = []
        for block in resp.content:
            if block.type != "tool_use":
                continue
            fn = TOOL_FUNCS[block.name]
            observacion = fn(**block.input)        # ejecuta NUESTRO código Python
            if verbose:
                print(f"  🔧 {block.name}({block.input}) → {observacion}")
            tool_results.append({
                "type": "tool_result",
                "tool_use_id": block.id,           # debe matchear el id del tool_use
                "content": str(observacion),
            })
        messages.append({"role": "user", "content": tool_results})

    # 4) TOPE: si llegamos acá, el agente no convergió → cortamos (anti loop infinito)
    return "⚠️ Se alcanzó el tope de iteraciones sin respuesta final."

print("Loop ReAct definido. Tope por defecto:", 6, "iteraciones.")
```

**Por qué este patrón y no un pipeline fijo.** En un *pipeline fijo* nosotros decidimos de antemano la secuencia de pasos (primero calculá, después buscá la fecha, después combiná). Funciona solo si conocemos la tarea exacta. El patrón **ReAct** invierte el control: el modelo **razona**, decide qué herramienta necesita, la **pide** vía `tool_use`, **observa** el `tool_result` y vuelve a razonar con esa nueva información. El mismo agente resuelve preguntas que ni anticipamos, porque la secuencia la arma él en tiempo de ejecución.

**El ciclo, paso a paso:**

1. **Razonar** — mandamos los `messages` al modelo con la lista de `tools`. El modelo responde texto    (si ya puede contestar) o uno o más bloques `tool_use` con el nombre y los argumentos.
2. **Actuar** — detectamos `stop_reason == "tool_use"`, buscamos la función Python por nombre y la    ejecutamos con los argumentos que pidió. Este es **nuestro** código, no del modelo.
3. **Observar** — devolvemos el resultado como un bloque `tool_result` (con el `tool_use_id` que    matchea el `tool_use`), y volvemos al paso 1.

**Por qué el tope de iteraciones es obligatorio.** El loop termina *naturalmente* cuando el modelo deja de pedir tools (`stop_reason != "tool_use"`). Pero un modelo confundido —o una tool que devuelve un error que el modelo no sabe interpretar— puede pedir herramientas indefinidamente, quemando tokens y plata. El `max_iteraciones` es un **circuit breaker**: garantiza terminación. No es un detalle de implementación, es un requisito de robustez de cualquier agente en producción.

## 🚀 4 · Una tarea con ≥2 herramientas

Le damos una pregunta que mezcla un cálculo y la fecha de hoy: para resolverla, el agente necesita usar dos tools distintas.

```python
# ───────────────────────────────────────────────────────────────────────────
# Tarea que NECESITA ≥2 herramientas: un cálculo y la fecha de hoy.
# Observá en la salida cómo el modelo pide 'calculadora' y 'fecha_hora'.
# ───────────────────────────────────────────────────────────────────────────
pregunta = "¿Cuánto es el 17% de 1.480, y qué día de la semana es hoy?"
respuesta = agente_react(pregunta)
print("\n" + "=" * 60)
print("PREGUNTA :", pregunta)
print("RESPUESTA:", respuesta)
```

### 📌 Qué vas a observar

Al correr la última celda, **no vas a ver un único texto**: vas a ver la **traza** del agente imprimiéndose iteración por iteración. Lo esperado para la pregunta *"¿Cuánto es el 17% de 1.480, y qué día de la semana es hoy?"*:

- **Iteración 1** — el modelo razona y, en lugar de inventar, **pide la tool** `calculadora` con `{'expresion': '17/100*1480'}` (o equivalente). Verás la línea `🔧 calculadora(...) → 251.6`. Puede pedir además `fecha_hora()` en el mismo turno (tool use en paralelo) o en una iteración aparte.
- **Observación** — nosotros ejecutamos esas funciones de Python y le devolvemos los resultados como `tool_result`. El modelo **no** calculó el 17%: lo hizo nuestra calculadora.
- **Iteración siguiente** — con las observaciones ya en contexto, el modelo deja de pedir tools (`stop_reason != "tool_use"`) y redacta la **respuesta final**, combinando ambos datos: *"El 17% de 1.480 es 251,6, y hoy es <día>."*

**Cómo interpretarlo.** La clave es que el modelo **no resolvió la tarea de un saque**: la descompuso, delegó cada parte en una herramienta determinista y recién entonces sintetizó. Esa traza —pedir tool, recibir resultado, concluir— es la firma del patrón ReAct, y es lo que distingue a un *agente* de un simple prompt. El número exacto de iteraciones puede variar entre corridas (el modelo puede agrupar las dos tools en un solo turno o pedirlas por separado); lo que **no** cambia es que la respuesta final se apoya en observaciones reales, no en una alucinación numérica.

## ✅ Conclusiones

- El patrón **ReAct** (razonar→actuar→observar) deja que el modelo elija herramientas dinámicamente, en vez de un pipeline fijo cableado por nosotros.
- El **function-calling** con JSON schema es el contrato: el modelo devuelve <code>tool_use</code> estructurado, nosotros ejecutamos código real y devolvemos <code>tool_result</code>.
- El **tope de iteraciones** es un requisito de robustez, no un detalle: sin él, un modelo confundido puede pedir tools indefinidamente.
