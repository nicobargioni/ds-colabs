### 🔍 Cómo leer este EDA

El EDA no es decorativo: **cada hallazgo condiciona una decisión de modelado**.

- **Faltantes** → definen la estrategia de imputación. Pocos nulos se imputan (mediana en numéricas, moda en categóricas); demasiados pueden justificar descartar la columna. **En este dataset no hay faltantes**: las 8.000 filas × 10 columnas están completas, así que la imputación es un seguro de robustez, no una necesidad.
- **Distribución / balance del target** → si las clases están desbalanceadas, la *accuracy* engaña; por eso más abajo evaluamos con **ROC-AUC / F1-macro**. En regresión, un target muy sesgado puede pedir transformaciones (log). Acá el target es `price`, fuertemente **sesgado a la derecha**: media ≈ 3.909 USD pero mediana 2.376 USD, mínimo 335 y máximo 18.823 USD. Esa cola larga (precios altos poco frecuentes) anticipa que los errores grandes vendrán de los diamantes caros.
- **Distribuciones y outliers** → sesgos fuertes y valores extremos favorecen escalado robusto y modelos de árboles (que no asumen normalidad). La regla IQR marca **6,90% de outliers en `price`** y 5,66% en `depth`, lo que refuerza la elección de un modelo basado en árboles.
- **Correlaciones** → variables muy correlacionadas aportan información redundante (afecta más a modelos lineales que a árboles) y pueden anticipar multicolinealidad. La correlación más fuerte con el precio es **`carat` = 0,92**, seguida por las dimensiones físicas **`y` = 0,89, `x` = 0,88 y `z` = 0,88**. Entre sí, esas dimensiones están casi perfectamente correlacionadas (**`x`–`y` = 0,999**) y `carat`–`x` = 0,975: hay multicolinealidad severa, que penaliza a la regresión lineal pero apenas afecta al bosque aleatorio. En cambio `table` (0,14) y `depth` (−0,04) casi no explican el precio.
- **Cardinalidad de categóricas** → una cardinalidad alta hace explotar el *one-hot encoding*; conviene vigilarla antes de preprocesar. Las tres categóricas son de cardinalidad baja y manejable: `cut` (5 niveles, dominado por *Ideal* con 3.166 filas), `color` (7 niveles, *G* el más común) y `clarity` (8 niveles, *SI1* el más común). El one-hot suma apenas ~17 columnas.

## 🧪 3 · Preprocesamiento

Split antes de transformar; imputación, escalado y encoding dentro de un `Pipeline` (ajustado solo con train → sin leakage).

### 🧪 Por qué este preprocesamiento

- **Primero separar, después transformar.** Si imputáramos o escaláramos con todo el dataset, el test "filtraría" información al entrenamiento (*data leakage*) y las métricas saldrían infladas. Por eso el split va **antes** y las transformaciones se ajustan **solo con train**, encapsuladas en un `Pipeline`.
- **Imputación:** mediana en numéricas (robusta a outliers), moda en categóricas.
- **Escalado** (`StandardScaler`): media 0, desvío 1. Crítico para modelos sensibles a la escala (lineales, basados en distancias); los árboles no lo necesitan, pero no se perjudican.
- **Encoding** (`OneHotEncoder`, `handle_unknown='ignore'`): categóricas → columnas binarias, tolerando categorías nuevas en test.
- **Split estratificado** (clasificación): preserva la proporción de clases en train y test.

```python
# 3.1 X / y y split train/test (estratificado si es clasificación)
TARGET = "price"
assert TARGET in df.columns, f'Target no encontrado: {TARGET}'
X = df.drop(columns=[TARGET]); y = df[TARGET]
strat = y if "regresion" == 'clasificacion' else None
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=SEED, stratify=strat)
print('train:', X_train.shape, '· test:', X_test.shape)
```

```python
# 3.2 Preprocesador (se ajusta SOLO con train dentro del Pipeline → sin leakage)
num_feats = X_train.select_dtypes(include='number').columns.tolist()
cat_feats = X_train.select_dtypes(exclude='number').columns.tolist()
numeric = Pipeline([('imp', SimpleImputer(strategy='median')), ('sc', StandardScaler())])
categ = Pipeline([('imp', SimpleImputer(strategy='most_frequent')),
                  ('oh', OneHotEncoder(handle_unknown='ignore', sparse_output=False))])
preprocess = ColumnTransformer([('num', numeric, num_feats), ('cat', categ, cat_feats)])
print(f'{len(num_feats)} numéricas + {len(cat_feats)} categóricas → preprocesador listo')
```

## 🧠 4 · Modelado

Baseline interpretable vs. ensemble, comparados con validación cruzada.

### 🧠 Qué modelos y por qué

1. **Regresión lineal** — *baseline* interpretable: asume relación lineal entre features y target.
2. **Random Forest Regressor** — *ensemble* que captura no linealidades e interacciones sin que tengamos que especificarlas.

**Hiperparámetros**: defaults sensatos (`n_estimators=300`, `random_state=42`), sin tuning todavía. **Validación cruzada 5-fold** con **R²** como métrica: una estimación promediada del poder explicativo, más confiable que un único split.

**Qué pasó**: el **Random Forest ganó con claridad**, con **R² = 0,973 ± 0,003** frente a **0,914 ± 0,012** de la regresión lineal, una mejora de **Δ +0,059** sobre el baseline. No fue un empate: la diferencia (≈6 puntos de varianza explicada) confirma que la relación entre las features y el precio **no es lineal** —tiene interacciones y curvatura, especialmente vía `carat` y las dimensiones físicas— y que el bosque las captura sin que tengamos que modelarlas a mano. La baja desviación del Random Forest (±0,003 vs. ±0,012 del lineal) también indica que su rendimiento es **más estable** entre folds.

```python
# 4 · Candidatos (baseline + ensemble) con validación cruzada
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
cv = KFold(n_splits=5, shuffle=True, random_state=SEED)
cands = {'LinReg (baseline)': LinearRegression(),
         'RandomForest': RandomForestRegressor(n_estimators=300, random_state=SEED, n_jobs=-1)}
results = {}
for name, reg in cands.items():
    pipe = Pipeline([('prep', preprocess), ('reg', reg)])
    sc = cross_val_score(pipe, X_train, y_train, cv=cv, scoring='r2', n_jobs=-1)
    results[name] = sc; print(f'{name:22s} R2: {sc.mean():.4f} ± {sc.std():.4f}')
best_name = max(results, key=lambda k: results[k].mean()); _b = next(k for k in results if 'baseline' in k.lower())
print(f'\nMejor modelo: {best_name} = {results[best_name].mean():.3f} · baseline {results[_b].mean():.3f} (Δ {results[best_name].mean()-results[_b].mean():+.3f})')
best = Pipeline([('prep', preprocess), ('reg', cands[best_name])]).fit(X_train, y_train)
```

## 📊 5 · Evaluación

Desempeño sobre datos nunca vistos (set de test).

```python
# 5 · Evaluación en test (datos nunca vistos)
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
y_pred = best.predict(X_test)
print(f'R2  : {r2_score(y_test, y_pred):.4f}')
print(f'MAE : {mean_absolute_error(y_test, y_pred):.4f}')
print(f'RMSE: {mean_squared_error(y_test, y_pred)**0.5:.4f}')
lims = [float(min(y_test.min(), y_pred.min())), float(max(y_test.max(), y_pred.max()))]
fig = px.scatter(x=y_test, y=y_pred, opacity=.5, labels={'x': 'real', 'y': 'predicho'},
                 title='Predicho vs. real')
fig.update_traces(marker_color=PRIMARY)
fig.add_trace(go.Scatter(x=lims, y=lims, mode='lines', line=dict(dash='dash', color=ACCENT), showlegend=False))
fig.show()
```

### 📌 Lectura de resultados

El Random Forest cierra el flujo con un **R² de 0,972 en test**, prácticamente idéntico al **0,973 de validación cruzada**: el modelo generaliza sin signos de sobreajuste y mejora **+0,059** sobre la regresión lineal (0,914), lo que prueba que la relación precio–features es no lineal y está bien capturada por el ensemble. En términos de negocio, las predicciones se desvían en promedio **334 USD (MAE)**, alrededor del 8,5% sobre un precio medio de ~3.909 USD. El **RMSE de 656 USD —casi el doble del MAE—** revela que el error no se reparte parejo: los diamantes caros de la cola derecha (`price` llega a 18.823 USD) concentran los desvíos mayores, tal como anticipaba el 6,9% de outliers detectado en el EDA. El motor predictivo es esencialmente `carat` (correlación 0,92 con el precio) junto a las dimensiones físicas `x`, `y` y `z` (≈0,88), mientras que `depth` y `table` aportan poco. El próximo salto de precisión vendría de modelar `log(price)` y ajustar hiperparámetros para domar esa cola de precios altos.

### 📊 Cómo interpretar estas métricas

- **R²**: proporción de varianza explicada (1 = perfecto; 0 = no mejor que predecir la media; negativo = peor que la media).
- **MAE**: error absoluto medio, en las **unidades del target** (lectura directa).
- **RMSE**: penaliza más los errores grandes; si **RMSE ≫ MAE**, hay outliers con error alto.
- **Predicho vs. real**: los puntos deberían alinearse sobre la diagonal; desvíos sistemáticos = sesgo del modelo.

**Lectura de los números reales de este test**:

- **R² = 0,9719** sobre 1.600 diamantes nunca vistos: el modelo explica el **97,2% de la varianza** del precio. Que casi coincida con el R² de validación cruzada (0,973) descarta sobreajuste —generaliza igual de bien fuera de entrenamiento.
- **MAE = 334 USD**: en promedio, la predicción se desvía unos 334 dólares del precio real. Sobre un precio medio de ~3.909 USD, es un **error relativo de ~8,5%**.
- **RMSE = 656 USD**, casi el **doble del MAE**. Esa brecha confirma lo que adelantaba el EDA: existe un grupo de **errores grandes**, concentrados en los diamantes caros (la cola derecha del precio), que el RMSE penaliza más fuerte. El grueso de las predicciones es preciso, pero los diamantes premium siguen siendo los más difíciles de tasar.

**Limitación**: sin tuning; probar regularización (Ridge/Lasso) y *feature engineering* (p. ej. modelar `log(price)` para domar la cola).
