### 🔍 Cómo leer este EDA

El EDA no es decorativo: **cada hallazgo condiciona una decisión de modelado**.

- **Faltantes** → definen la estrategia de imputación. Pocos nulos se imputan (mediana en numéricas, moda en categóricas); demasiados pueden justificar descartar la columna.
- **Distribución / balance del target** → si las clases están desbalanceadas, la *accuracy* engaña; por eso más abajo evaluamos con **ROC-AUC / F1-macro**. En regresión, un target muy sesgado puede pedir transformaciones (log).
- **Distribuciones y outliers** → sesgos fuertes y valores extremos favorecen escalado robusto y modelos de árboles (que no asumen normalidad).
- **Correlaciones** → variables muy correlacionadas aportan información redundante (afecta más a modelos lineales que a árboles) y pueden anticipar multicolinealidad.
- **Cardinalidad de categóricas** → una cardinalidad alta hace explotar el *one-hot encoding*; conviene vigilarla antes de preprocesar.

**Lectura concreta de este dataset (Auto MPG):**

- El DataFrame tiene **398 filas × 8 columnas**: 7 numéricas y 1 categórica (`origin`). El target `mpg` es continuo, así que esto es un problema de **regresión**, no de clasificación.
- **Faltantes muy acotados**: solo `horsepower` tiene nulos, **6 valores (1.51%)**. Es un volumen ínfimo, perfectamente imputable por mediana sin descartar la columna.
- **Distribución del target**: `mpg` va de **9.0 a 46.6**, con media **23.5** y mediana **23.0**. La cercanía media≈mediana indica una distribución bastante simétrica, sin sesgo fuerte que obligue a transformar con log.
- **Correlaciones con `mpg`**: las más fuertes son **weight = -0.83**, **displacement = -0.80**, **horsepower = -0.78** y **cylinders = -0.78**, todas negativas (autos más pesados y de mayor cilindrada consumen más, o sea menos millas por galón). `model_year` aporta señal positiva moderada (**+0.58**) y `acceleration` la más débil (**+0.42**). Las cuatro variables de "tamaño/potencia" están además muy correlacionadas entre sí → **multicolinealidad esperable**, que penaliza a los modelos lineales pero no a los árboles.
- **Outliers (regla IQR)**: muy pocos: `horsepower` 10 (2.51%), `acceleration` 7 (1.76%) y `mpg` 1 (0.25%). Nada que requiera limpieza agresiva.
- **Categórica `origin`**: cardinalidad baja (3 valores: usa 249, japan 79, europe 70), one-hot trivial.

## 🧪 3 · Preprocesamiento

Split antes de transformar; imputación, escalado y encoding dentro de un `Pipeline` (ajustado solo con train → sin leakage).

### 🧪 Por qué este preprocesamiento

- **Primero separar, después transformar.** Si imputáramos o escaláramos con todo el dataset, el test "filtraría" información al entrenamiento (*data leakage*) y las métricas saldrían infladas. Por eso el split va **antes** y las transformaciones se ajustan **solo con train**, encapsuladas en un `Pipeline`.
- **Imputación:** mediana en numéricas (robusta a outliers), moda en categóricas.
- **Escalado** (`StandardScaler`): media 0, desvío 1. Crítico para modelos sensibles a la escala (lineales, basados en distancias); los árboles no lo necesitan, pero no se perjudican.
- **Encoding** (`OneHotEncoder`, `handle_unknown='ignore'`): categóricas → columnas binarias, tolerando categorías nuevas en test.
- **Split estratificado** (clasificación): preserva la proporción de clases en train y test.

```python
# 3.1 X / y y split train/test (estratificado si es clasificación)
TARGET = "mpg"
assert TARGET in df.columns, f'Target no encontrado: {TARGET}'
X = df.drop(columns=[TARGET]); y = df[TARGET]
strat = y if "regresion" == 'clasificacion' else None
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=SEED, stratify=strat)
print('train:', X_train.shape, '· test:', X_test.shape)
```

```python
# 3.2 Preprocesador (se ajusta SOLO con train dentro del Pipeline → sin leakage)
num_feats = X_train.select_dtypes(include='number').columns.tolist()
cat_feats = X_train.select_dtypes(exclude='number').columns.tolist()
numeric = Pipeline([('imp', SimpleImputer(strategy='median')), ('sc', StandardScaler())])
categ = Pipeline([('imp', SimpleImputer(strategy='most_frequent')),
                  ('oh', OneHotEncoder(handle_unknown='ignore', sparse_output=False))])
preprocess = ColumnTransformer([('num', numeric, num_feats), ('cat', categ, cat_feats)])
print(f'{len(num_feats)} numéricas + {len(cat_feats)} categóricas → preprocesador listo')
```

## 🧠 4 · Modelado

Baseline interpretable vs. ensemble, comparados con validación cruzada.

### 🧠 Qué modelos y por qué

1. **Regresión lineal** — *baseline* interpretable: asume relación lineal entre features y target.
2. **Random Forest Regressor** — *ensemble* que captura no linealidades e interacciones sin que tengamos que especificarlas.

**Hiperparámetros**: defaults sensatos (`n_estimators=300`, `random_state=42`), sin tuning todavía. **Validación cruzada 5-fold** con **R²** como métrica: una estimación promediada del poder explicativo, más confiable que un único split.

**Qué pasó en la validación cruzada (5-fold, train):**

- **RandomForest ganó** con **R² = 0.854 ± 0.027**, contra **R² = 0.803 ± 0.019** de la regresión lineal: una mejora de **Δ +0.051** sobre el baseline.
- No es un empate: el bosque captura la **no linealidad** de la relación consumo↔peso/cilindrada y maneja sin problema la multicolinealidad detectada en el EDA, algo que limita al modelo lineal. La ventaja de ~5 puntos de R² confirma que el problema **no es puramente lineal** y justifica el ensemble.
- Aun así, el baseline lineal ya explica el 80% de la varianza: las relaciones dominantes son fuertes y casi monótonas, por eso ambos modelos rinden alto.

```python
# 4 · Candidatos (baseline + ensemble) con validación cruzada
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
cv = KFold(n_splits=5, shuffle=True, random_state=SEED)
cands = {'LinReg (baseline)': LinearRegression(),
         'RandomForest': RandomForestRegressor(n_estimators=300, random_state=SEED, n_jobs=-1)}
results = {}
for name, reg in cands.items():
    pipe = Pipeline([('prep', preprocess), ('reg', reg)])
    sc = cross_val_score(pipe, X_train, y_train, cv=cv, scoring='r2', n_jobs=-1)
    results[name] = sc; print(f'{name:22s} R2: {sc.mean():.4f} ± {sc.std():.4f}')
best_name = max(results, key=lambda k: results[k].mean()); _b = next(k for k in results if 'baseline' in k.lower())
print(f'\nMejor modelo: {best_name} = {results[best_name].mean():.3f} · baseline {results[_b].mean():.3f} (Δ {results[best_name].mean()-results[_b].mean():+.3f})')
best = Pipeline([('prep', preprocess), ('reg', cands[best_name])]).fit(X_train, y_train)
```

## 📊 5 · Evaluación

Desempeño sobre datos nunca vistos (set de test).

```python
# 5 · Evaluación en test (datos nunca vistos)
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
y_pred = best.predict(X_test)
print(f'R2  : {r2_score(y_test, y_pred):.4f}')
print(f'MAE : {mean_absolute_error(y_test, y_pred):.4f}')
print(f'RMSE: {mean_squared_error(y_test, y_pred)**0.5:.4f}')
lims = [float(min(y_test.min(), y_pred.min())), float(max(y_test.max(), y_pred.max()))]
fig = px.scatter(x=y_test, y=y_pred, opacity=.5, labels={'x': 'real', 'y': 'predicho'},
                 title='Predicho vs. real')
fig.update_traces(marker_color=PRIMARY)
fig.add_trace(go.Scatter(x=lims, y=lims, mode='lines', line=dict(dash='dash', color=ACCENT), showlegend=False))
fig.show()
```

### 📌 Lectura de resultados

El **RandomForest** (modelo ganador en CV) alcanza sobre el conjunto de test un **R² de 0.912**, explicando el **91% de la varianza** del consumo en datos nunca vistos. El **MAE de 1.60 mpg** implica un error medio de apenas ~7% respecto de la media del target (23.5 mpg), un nivel de precisión muy bueno para una predicción de consumo. El **RMSE de 2.18 mpg** supera al MAE en ~36%, lo que indica que el grueso del error es bajo y uniforme, pero hay **unos pocos vehículos atípicos** —consistentes con los outliers de `horsepower` detectados en el EDA— que arrastran el error cuadrático hacia arriba. La mejora de **+0.051 de R²** del bosque sobre el baseline lineal en validación cruzada confirma que las interacciones y no linealidades entre peso, cilindrada y año de modelo aportan señal real. En conjunto, el modelo **generaliza con solidez** y el residual remanente se concentra en casos extremos, no en un sesgo sistemático.

### 📊 Cómo interpretar estas métricas

- **R²**: proporción de varianza explicada (1 = perfecto; 0 = no mejor que predecir la media; negativo = peor que la media).
- **MAE**: error absoluto medio, en las **unidades del target** (lectura directa).
- **RMSE**: penaliza más los errores grandes; si **RMSE ≫ MAE**, hay outliers con error alto.
- **Predicho vs. real**: los puntos deberían alinearse sobre la diagonal; desvíos sistemáticos = sesgo del modelo.

**Interpretación de los números reales (RandomForest sobre el test, datos nunca vistos):**

- **R² = 0.912** → el modelo explica el **91% de la varianza** del consumo en test. Es incluso mejor que el R² de validación cruzada (0.854), señal de un test split favorable pero sin indicios de sobreajuste grave.
- **MAE = 1.60** → en promedio el modelo se equivoca por **~1.6 mpg**. Sobre un target con media 23.5 mpg, es un error relativo de apenas **~7%**: muy preciso para uso práctico.
- **RMSE = 2.18** vs MAE 1.60 → el RMSE es ~36% mayor que el MAE, lo que indica que **existen unos pocos autos con error bastante más alto** que la media (coherente con los outliers de `horsepower` vistos en el EDA), pero no es una brecha dramática.
- **Qué se infiere**: las variables de tamaño/potencia (weight, displacement, horsepower) más `model_year` son predictores potentes y estables del consumo; el modelo generaliza bien y el error residual es pequeño y se concentra en casos atípicos.

**Limitación**: sin tuning; probar regularización (Ridge/Lasso) y *feature engineering*.
