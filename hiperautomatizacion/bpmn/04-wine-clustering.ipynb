### 🔍 Cómo leer este EDA

El EDA no es decorativo: **cada hallazgo condiciona una decisión de modelado**. Conceptos generales primero, y debajo la lectura concreta de **este** dataset.

- **Faltantes** → definen la estrategia de imputación. Pocos nulos se imputan (mediana en numéricas, moda en categóricas); demasiados pueden justificar descartar la columna.
- **Distribución / balance del target** → si las clases están desbalanceadas, la *accuracy* engaña. Acá el problema es **no supervisado**: no hay target, así que la calidad se mide por cohesión/separación (silueta), no por F1/ROC-AUC.
- **Distribuciones y outliers** → sesgos fuertes y valores extremos favorecen escalado robusto y modelos que no asumen normalidad.
- **Correlaciones** → variables muy correlacionadas aportan información redundante y anticipan multicolinealidad.
- **Cardinalidad de categóricas** → una cardinalidad alta hace explotar el *one-hot encoding*.

**Lectura concreta de este dataset (Wine UCI):**
- Forma: **178 filas × 13 columnas**, las **13 numéricas** (`float64`) y **0 categóricas**. Es un dataset chico y limpio.

- Sin target supervisado en esta etapa: la columna `target` original se descartó a propósito para tratarlo como problema de **descubrimiento de grupos**.

## 🧪 3 · Preprocesamiento

Escalado y encoding sobre todo el conjunto (problema no supervisado, sin target).

### 🧪 Por qué este preprocesamiento

En un problema **no supervisado** no hay target ni split: el objetivo es describir la estructura de *todos* los datos.

- **Escalado** (`StandardScaler`): imprescindible acá, porque tanto K-Means como Isolation Forest se basan en **distancias**; sin escalar, las variables de mayor magnitud dominan artificialmente.
- **Encoding** de categóricas a binarias para que entren en el cálculo de distancias.
- Todo se ajusta sobre el conjunto completo (no hay fuga posible porque no hay test que predecir).

```python
# 3 · Preprocesamiento (no supervisado): escalado + encoding sobre todo X
num_feats = df.select_dtypes(include='number').columns.tolist()
cat_feats = df.select_dtypes(exclude='number').columns.tolist()
numeric = Pipeline([('imp', SimpleImputer(strategy='median')), ('sc', StandardScaler())])
categ = Pipeline([('imp', SimpleImputer(strategy='most_frequent')),
                  ('oh', OneHotEncoder(handle_unknown='ignore', sparse_output=False))])
preprocess = ColumnTransformer([('num', numeric, num_feats), ('cat', categ, cat_feats)])
Xp = preprocess.fit_transform(df)
print('Matriz preprocesada:', Xp.shape)
```

## 🧠 4 · Modelado

K-Means con selección de k (codo + silueta).

### 🧠 Por qué K-Means y cómo elegir *k*

**K-Means** agrupa minimizando la distancia de cada punto a su centroide. Es rápido e interpretable, ideal como primer mapeo de la estructura.

Como *k* (cantidad de grupos) no se conoce de antemano, lo elegimos con dos criterios:
- **Método del codo (inercia)**: se busca el quiebre donde sumar clusters deja de reducir mucho la inercia.
- **Coeficiente de silueta**: mide cohesión vs. separación; elegimos el *k* que lo maximiza.

**Qué pasó al ejecutar:** barriendo *k* de 2 a 8, la silueta marcó su **máximo en k=3 (silueta ≈ 0.285)**, por encima de k=2 (0.259) y k=4 (0.260), y cayendo claramente a partir de ahí (k=8 baja a 0.157). La inercia, por su lado, hace su mayor caída entre k=2 (1659) y k=3 (1278) y luego se aplana — el codo coincide con la silueta. **Ambos criterios apuntan al mismo k=3**, una señal robusta de que tres grupos describen la estructura natural del dataset (que, además, son exactamente las tres variedades de vino reales).

```python
# 4 · K-Means: elegir k con el codo (inercia) y la silueta
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
ks = list(range(2, 9)); inertias = []; sils = []
for k in ks:
    km = KMeans(n_clusters=k, random_state=SEED, n_init=10).fit(Xp)
    inertias.append(km.inertia_); sils.append(silhouette_score(Xp, km.labels_))
fig = sp.make_subplots(rows=1, cols=2, subplot_titles=('Codo (inercia)', 'Silueta'))
fig.add_trace(go.Scatter(x=ks, y=inertias, mode='lines+markers', line=dict(color=PRIMARY)), row=1, col=1)
fig.add_trace(go.Scatter(x=ks, y=sils, mode='lines+markers', line=dict(color=ACCENT)), row=1, col=2)
fig.update_layout(height=360, showlegend=False, template='plotly_dark'); fig.show()
best_k = ks[int(np.argmax(sils))]
print('k elegido por silueta:', best_k)
km = KMeans(n_clusters=best_k, random_state=SEED, n_init=10).fit(Xp)
df['cluster'] = km.labels_
```

## 📊 5 · Evaluación

Interpretación de los grupos / anomalías (proyección PCA 2D).

```python
# 5 · Interpretación: clusters proyectados en 2D (PCA)
from sklearn.decomposition import PCA
pca = PCA(n_components=2, random_state=SEED); coords = pca.fit_transform(Xp)
px.scatter(x=coords[:, 0], y=coords[:, 1], color=df['cluster'].astype(str),
           labels={'x': 'PC1', 'y': 'PC2', 'color': 'cluster'},
           title=f'Clusters (PCA 2D) · varianza explicada {pca.explained_variance_ratio_.sum():.0%}').show()
display(df.groupby('cluster').size().rename('n').to_frame())
```

El resultado es un clustering **balanceado de 65/62/51 vinos**, sin grupos degenerados. La silueta moderada (lejos de 1) y el hecho de que la proyección PCA 2D solo capture el **55.4% de la varianza** (36.2% + 19.2%) indican que los grupos son reales pero con bordes difusos y estructura repartida en más dimensiones de las que se pueden dibujar. El dato más fuerte: al comparar los 3 clusters descubiertos contra las 3 variedades reales de vino que se habían ocultado, el **ARI = 0.897** — es decir, K-Means reconstruyó la clasificación verdadera con ~90% de concordancia partiendo solo de las features químicas estandarizadas. Se infiere que la composición química (con el bloque fenoles/flavonoides como eje principal) separa las variedades de forma casi natural, y que el preprocesamiento de escalado fue determinante para que la distancia euclídea capturara esa señal.

### 📊 Cómo interpretar el clustering

- **Codo / silueta**: guiaron la elección de *k*. El máximo obtenido fue **silueta ≈ 0.285 en k=3**. - **PCA 2D**: las dos primeras componentes explican **55.4% de la varianza** (PC1 = 36.2%, PC2 = 19.2%). Es una *proyección* para visualizar, no la separación real: el ~45% de varianza restante vive en dimensiones que el gráfico no muestra, así que grupos que parezcan tocarse en el plano pueden estar más separados en alta dimensión.
- **Tamaño de los grupos**: el reparto quedó **balanceado — 65 / 62 / 51 vinos** (clusters 0 / 2 / 1), sin grupos residuales ni un cluster que se trague a la mayoría. Eso da confianza en que K-Means encontró estructura real y no ruido.
- **Qué inferir**: caracterizá cada cluster mirando las **medias de las variables por grupo** (perfiles). Dado el bloque fenoles/flavonoides tan correlacionado del EDA, esas variables probablemente sean las que más separan los grupos.

**Limitación**: K-Means asume grupos esféricos y de tamaño similar; para formas complejas conviene DBSCAN o Gaussian Mixtures.

## ✅ Conclusiones

- **Wine (UCI)** resultó un dataset chico y limpio: **178 × 13**, todas numéricas, **0 faltantes** y casi sin outliers (máximo 4 = 2.25% en `magnesium`, `alcalinity_of_ash` y `color_intensity`).
- El EDA reveló un bloque químico muy redundante: **`total_phenols`–`flavanoids` = 0.86** es la correlación dominante, lo que anticipa que pocos ejes concentran la información.
- **K-Means con codo + silueta convergió en k=3** (silueta ≈ 0.285), con grupos balanceados de **65/62/51** vinos — ni el codo ni la silueta dejaron ambigüedad.
- La proyección **PCA 2D explica 55.4%** de la varianza; la separación visual es parcial porque casi la mitad de la información vive fuera del plano.
- **Validación contra la verdad oculta: ARI = 0.897** frente a las 3 variedades reales — el clustering no supervisado recuperó casi toda la estructura conocida, evidencia de que la firma química discrimina las variedades.
- El preprocesamiento vive dentro de un **Pipeline** (escalado obligatorio dado el rango de `proline` vs `hue`) → sin data leakage.
