## 🔍 2 · Análisis exploratorio (EDA de texto)

Balance de clases, longitud de los textos y vocabulario más frecuente.

```python
# 2 · EDA de texto
print('Documentos:', len(df))
display(df['label'].value_counts().rename('n').to_frame())
df['_len'] = df['text'].astype(str).str.len()
display(df['_len'].describe().to_frame().T)
px.histogram(df, x='_len', nbins=50, color='label',
             title='Longitud de los textos (caracteres) por clase', labels={'_len':'caracteres'}).show()
from sklearn.feature_extraction.text import CountVectorizer
cvz = CountVectorizer(max_features=20, stop_words='english')
m = cvz.fit_transform(df['text'].astype(str))
freq = pd.Series(m.sum(axis=0).A1, index=cvz.get_feature_names_out()).sort_values()
px.bar(x=freq.values, y=freq.index, orientation='h',
       title='Tokens más frecuentes (sin stopwords)').show()
```

### 🔍 Cómo leer este EDA de texto

- **Balance de clases**: si una clase domina, la *accuracy* engaña → evaluamos con **F1-macro**. En este corpus el desbalance es marcado: **4825 ham (86.6%) vs 747 spam (13.4%)**, casi 6.5 a 1. Un clasificador que dijera "todo ham" acertaría el 86.6% y aun así sería inútil para detectar spam.
- **Longitud de los textos**: la distribución de caracteres es muy asimétrica (mediana **62**, media **80.5**, máximo **910**, con cola larga a la derecha). El histograma por clase muestra que los **spam tienden a ser más largos** (promociones, premios, enlaces), mientras que los ham concentran su masa en mensajes cortos — una señal aprovechable por el modelo.
- **Tokens frecuentes**: tras quitar *stopwords*, los términos más comunes del corpus (`gt`, `lt`, `ok`, `ur`, `just`) son mayormente conversacionales/ham; el vocabulario discriminativo del spam (free, win, txt, claim, prize) aparece más abajo pero pesa fuerte vía IDF.
- Esto guía la vectorización: **TF-IDF** pondera términos frecuentes en un documento pero raros en el corpus (los más discriminativos), exactamente lo que separa el lenguaje promocional del spam del habla cotidiana.

## 🧪 3 · Representación del texto

TF-IDF dentro del `Pipeline` (el IDF se ajusta solo con train → sin leakage).

### 🧠 Por qué TF-IDF + modelos lineales

Para texto, **TF-IDF + un modelo lineal** es un baseline fortísimo y rápido (a menudo difícil de superar sin transformers):

1. **TF-IDF** (`ngram_range=(1,2)`, `sublinear_tf`, `stop_words='english'`): representa cada documento por la importancia de sus términos y bigramas.
2. **Regresión logística** vs **LinearSVC**: ambos lineales sobre el espacio TF-IDF; SVM lineal suele rendir muy bien en texto disperso de alta dimensión.

El vectorizador va **dentro del `Pipeline`** → el IDF se ajusta solo con train (sin leakage). Validación cruzada 5-fold con **F1-macro** (tolera desbalance).

**Qué pasó al entrenar**: no hubo empate. **LinearSVC ganó con claridad: F1-macro 0.9682 ± 0.0045 vs 0.8694 ± 0.0114 de la LogReg baseline (Δ +0.099)**. La diferencia (~10 puntos) es muy superior al desvío de los folds, así que es estadísticamente sólida: en texto disperso de alta dimensión el margen máximo de SVM separa mejor las clases minoritarias que la regresión logística, que sufre más bajo el desbalance 6.5:1. Confirma la intuición de que el SVM lineal es el caballo de batalla para este tipo de problema.

```python
# 3 · Representación del texto
# El texto se vectoriza con TF-IDF DENTRO del Pipeline del modelo, así el
# vocabulario y los pesos IDF se ajustan solo con train (sin data leakage).
print('Listo: la vectorización TF-IDF se hace dentro del Pipeline (sección 4).')
```

## 🧠 4 · Modelado

TF-IDF + modelos lineales (LogReg vs LinearSVC) con validación cruzada.

```python
# 4 · TF-IDF + modelos lineales con validación cruzada
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
X = df['text'].astype(str); y = df['label']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=SEED, stratify=y)
def mk(clf):
    return Pipeline([('tfidf', TfidfVectorizer(max_features=20000, ngram_range=(1,2),
                      stop_words='english', sublinear_tf=True)), ('clf', clf)])
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=SEED)
cands = {'LogReg (baseline)': mk(LogisticRegression(max_iter=1000, random_state=SEED)),
         'LinearSVC': mk(LinearSVC(random_state=SEED))}
results = {}
for name, pipe in cands.items():
    sc = cross_val_score(pipe, X_train, y_train, cv=cv, scoring='f1_macro', n_jobs=-1)
    results[name] = sc; print(f'{name:20s} F1-macro: {sc.mean():.4f} ± {sc.std():.4f}')
best_name = max(results, key=lambda k: results[k].mean()); _b = next(k for k in results if 'baseline' in k.lower())
print(f'\nMejor modelo: {best_name} = {results[best_name].mean():.3f} · baseline {results[_b].mean():.3f} (Δ {results[best_name].mean()-results[_b].mean():+.3f})')
best = cands[best_name].fit(X_train, y_train)
```

## 📊 5 · Evaluación

Desempeño sobre textos no vistos (set de test).

```python
# 5 · Evaluación en test
from sklearn.metrics import classification_report, confusion_matrix
y_pred = best.predict(X_test)
print(classification_report(y_test, y_pred))
labels = sorted(map(str, y.unique()))
cm = confusion_matrix(y_test, y_pred)
px.imshow(cm, x=labels, y=labels, text_auto=True, color_continuous_scale='Blues',
          labels=dict(x='predicho', y='real', color='n'), title='Matriz de confusión').show()
```

### 📌 Lectura de resultados

Sobre el set de test (1115 SMS no vistos), **LinearSVC alcanza accuracy 0.98 y F1-macro 0.96**, en línea con lo observado en validación cruzada (0.9682), señal de que **no hubo sobreajuste**: el desempeño se mantiene al pasar de train a datos nuevos. La matriz de confusión muestra **961 de 966 ham** y **134 de 149 spam** correctamente clasificados; el error se concentra en **15 spams que se filtran como ham** (falsos negativos) frente a apenas **5 ham marcados como spam**. Esto da un perfil de **alta precision en spam (0.96) con recall algo menor (0.90)**, exactamente lo que conviene a un filtro: rara vez molesta al usuario marcando mensajes legítimos, a cambio de dejar escapar 1 de cada 10 spams. De aquí se infiere que el cuello de botella no es la precisión sino la **cobertura del spam**: la palanca de mejora más directa es elevar el recall de la clase minoritaria (umbral de decisión, `class_weight='balanced'` o ajuste de hiperparámetros) sin sacrificar la precisión ya lograda.

### 📊 Cómo interpretar estas métricas

- **F1-macro**: promedia el F1 por clase sin que la mayoritaria domine — la métrica honesta cuando hay desbalance. En test el modelo final (LinearSVC) alcanza **F1-macro 0.96** y **accuracy 0.98** sobre 1115 mensajes no vistos.
- **Matriz de confusión**: de **966 ham**, clasifica bien **961** y confunde solo **5** como spam (falsos positivos); de **149 spam**, detecta **134** y deja escapar **15** como ham (falsos negativos). El error dominante es el **spam que se filtra como ham**, no al revés.
- **Precision vs Recall por clase**: para spam la **precision es 0.96** (cuando marca spam casi siempre acierta) pero el **recall es 0.90** — se le escapa 1 de cada 10 spams. En la clase mayoritaria ham todo es casi perfecto (precision 0.98, recall 0.99, F1 0.99). Para un filtro anti-spam este perfil es deseable: prioriza no marcar mensajes legítimos como spam (alta precision en spam) a costa de dejar pasar algún spam ocasional.

**Limitación**: TF-IDF ignora el orden y el contexto; para problemas donde la semántica fina importa, el paso siguiente son **embeddings** (sentence-transformers) o *fine-tuning* de un transformer.

- Próximos pasos: subir el **recall de spam** (ajuste de umbral / `class_weight`, hiperparámetros del SVM) y, si la semántica fina lo justifica, evaluar **embeddings o un transformer fine-tuneado**.
