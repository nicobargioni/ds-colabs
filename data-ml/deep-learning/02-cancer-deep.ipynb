### 🔍 Cómo leer este EDA

El EDA no es decorativo: **cada hallazgo condiciona una decisión de modelado**. Esto es lo que muestran los datos de **este** dataset (569 muestras × 30 features de imágenes → benigno/maligno):

- **Faltantes** → definen la estrategia de imputación. Acá **no hay un solo faltante** (las 31 columnas tienen 569 valores no nulos), así que la imputación por mediana del Pipeline nunca se activa: queda como red de seguridad, no como necesidad real.
- **Distribución / balance del target** → si las clases están desbalanceadas, la *accuracy* engaña; por eso evaluamos con **ROC-AUC / F1-macro**. El target está **moderadamente desbalanceado**: clase 1 (benigno) = 62.7% (357 casos) contra clase 0 (maligno) = 37.3% (212 casos), con un **ratio mín/máx de 0.59**. No es severo, pero confirma que mirar solo accuracy sería tramposo.
- **Distribuciones y outliers** → las 30 features son numéricas y de escalas muy dispares (`mean area` ~en miles vs `mean smoothness` ~0.1). Esa diferencia de magnitud hace **imprescindible el `StandardScaler`** para el modelo lineal, que es sensible a la escala.
- **Correlaciones** → variables muy correlacionadas aportan información redundante y anticipan multicolinealidad. Acá la señal predictiva es altísima: las más correlacionadas con el target son **`worst concave points` (r = −0.79)**, `worst perimeter` (−0.78) y `mean concave points` (−0.78). Además hay **multicolinealidad fuerte entre features**: `mean radius` ~ `mean perimeter` = 0.998 y `mean area` ~ `mean radius` = 0.987 (radio, perímetro y área miden esencialmente lo mismo).
- **Cardinalidad de categóricas** → no aplica: las 31 columnas son numéricas (30 `float64` + el target `int64`), 0 categóricas, así que el `OneHotEncoder` del Pipeline queda inactivo.


## 🧪 3 · Preprocesamiento

Split antes de transformar; imputación, escalado y encoding dentro de un `Pipeline` (ajustado solo con train → sin leakage).

### 🧪 Por qué este preprocesamiento

- **Primero separar, después transformar.** Si imputáramos o escaláramos con todo el dataset, el test "filtraría" información al entrenamiento (*data leakage*) y las métricas saldrían infladas. Por eso el split va **antes** y las transformaciones se ajustan **solo con train**, encapsuladas en un `Pipeline`.
- **Imputación:** mediana en numéricas (robusta a outliers), moda en categóricas.
- **Escalado** (`StandardScaler`): media 0, desvío 1. Crítico para modelos sensibles a la escala (lineales, basados en distancias); los árboles no lo necesitan, pero no se perjudican.
- **Encoding** (`OneHotEncoder`, `handle_unknown='ignore'`): categóricas → columnas binarias, tolerando categorías nuevas en test.
- **Split estratificado** (clasificación): preserva la proporción de clases en train y test.

```python
# 3.1 X / y y split train/test (estratificado si es clasificación)
TARGET = "target"
assert TARGET in df.columns, f'Target no encontrado: {TARGET}'
X = df.drop(columns=[TARGET]); y = df[TARGET]
strat = y if "deep" == 'clasificacion' else None
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=SEED, stratify=strat)
print('train:', X_train.shape, '· test:', X_test.shape)
```

```python
# 3.2 Preprocesador (se ajusta SOLO con train dentro del Pipeline → sin leakage)
num_feats = X_train.select_dtypes(include='number').columns.tolist()
cat_feats = X_train.select_dtypes(exclude='number').columns.tolist()
numeric = Pipeline([('imp', SimpleImputer(strategy='median')), ('sc', StandardScaler())])
categ = Pipeline([('imp', SimpleImputer(strategy='most_frequent')),
                  ('oh', OneHotEncoder(handle_unknown='ignore', sparse_output=False))])
preprocess = ColumnTransformer([('num', numeric, num_feats), ('cat', categ, cat_feats)])
print(f'{len(num_feats)} numéricas + {len(cat_feats)} categóricas → preprocesador listo')
```

## 🧠 4 · Modelado

Baseline interpretable vs. ensemble, comparados con validación cruzada.

### 🧠 Por qué una red neuronal acá

1. **Regresión logística** — *baseline* lineal, para tener una vara honesta.
2. **MLP (perceptrón multicapa)** — red neuronal con capas ocultas `(64, 32)`: aprende representaciones no lineales de los datos tabulares.

**Decisiones de la red**: `early_stopping=True` corta el entrenamiento cuando la validación deja de mejorar (evita sobreajuste); `max_iter=400` es el techo de épocas; las capas `(64,32)` son una arquitectura modesta acorde al tamaño del dataset (redes grandes sobreajustan en pocos datos). Se compara contra el baseline con **validación cruzada** — una red solo se justifica si **supera** al modelo simple.

**Qué pasó al ejecutar**: el baseline lineal **ganó**. En 5-fold estratificado, la **Regresión logística logró ROC-AUC 0.995 ± 0.006**, por encima del **MLP, que quedó en 0.983 ± 0.009** (~0.012 por debajo). El delta del mejor modelo sobre el baseline fue **+0.000**: la red neuronal no aportó nada. La lectura es clara — **el problema es esencialmente lineal**: las clases benigno/maligno son casi separables con una frontera lineal sobre las features escaladas (las correlaciones fuertes del EDA ya lo anticipaban), así que la capacidad no lineal del MLP solo agrega varianza sin mejorar la señal. Es el caso de libro donde la red **no se justifica**.


```python
# 4 · Red neuronal (MLP) vs. baseline lineal — validación cruzada
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=SEED)
scoring = 'roc_auc' if y.nunique()==2 else 'f1_macro'
cands = {'LogReg (baseline)': LogisticRegression(max_iter=1000, random_state=SEED),
         'MLP (red neuronal)': MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=400,
                                             early_stopping=True, random_state=SEED)}
results = {}
for name, clf in cands.items():
    pipe = Pipeline([('prep', preprocess), ('clf', clf)])
    sc = cross_val_score(pipe, X_train, y_train, cv=cv, scoring=scoring, n_jobs=-1)
    results[name] = sc; print(f'{name:22s} {scoring}: {sc.mean():.4f} ± {sc.std():.4f}')
best_name = max(results, key=lambda k: results[k].mean()); _b = next(k for k in results if 'baseline' in k.lower())
print(f'\nMejor modelo: {best_name} = {results[best_name].mean():.3f} · baseline {results[_b].mean():.3f} (Δ {results[best_name].mean()-results[_b].mean():+.3f})')
best = Pipeline([('prep', preprocess), ('clf', cands[best_name])]).fit(X_train, y_train)
```

## 📊 5 · Evaluación

Desempeño sobre datos nunca vistos (set de test).

```python
# 5 · Evaluación en test (datos nunca vistos)
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
y_pred = best.predict(X_test)
print(classification_report(y_test, y_pred))
labels = sorted(map(str, y.unique()))
cm = confusion_matrix(y_test, y_pred)
px.imshow(cm, x=labels, y=labels, text_auto=True, color_continuous_scale='Blues',
          labels=dict(x='predicho', y='real', color='n'), title='Matriz de confusión').show()
if y.nunique() == 2 and hasattr(best, 'predict_proba'):
    pos = sorted(y.unique())[-1]
    proba = best.predict_proba(X_test)[:, list(best.classes_).index(pos)]
    fpr, tpr, _ = roc_curve((y_test == pos).astype(int), proba)
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=fpr, y=tpr, mode='lines', line=dict(color=ACCENT, width=3),
                             name=f'AUC = {auc(fpr, tpr):.3f}', fill='tozeroy'))
    fig.add_trace(go.Scatter(x=[0, 1], y=[0, 1], mode='lines',
                             line=dict(dash='dash', color='gray'), showlegend=False))
    fig.update_layout(title='Curva ROC', xaxis_title='FPR', yaxis_title='TPR', template='plotly_dark')
    fig.show()
```

### 📌 Lectura de resultados

Sobre los **114 casos de test nunca vistos**, el modelo final (regresión logística) acertó el **97% (accuracy 0.97)** con un **ROC-AUC de 0.997**, casi un ranking perfecto entre benigno y maligno. El desempeño por clase es parejo: F1 de 0.96 en malignos (clase 0) y 0.98 en benignos (clase 1), sin que el desbalance 37/63 distorsione el resultado. El único error relevante son ~2 tumores malignos etiquetados como benignos (recall maligno 0.95): pocos en términos absolutos, pero el tipo de error más costoso en un contexto médico. Que el baseline lineal igualara y superara al MLP en validación cruzada (0.995 vs 0.983) confirma que la frontera entre clases es prácticamente lineal sobre las features escaladas. Conclusión cuantitativa: hay señal sobrada y un modelo simple basta; el margen de mejora no está en agregar capacidad no lineal sino en **ajustar el umbral para reducir falsos negativos**.


### 📊 Cómo interpretar estas métricas

- **Matriz de confusión**: la diagonal son aciertos; fuera de ella, errores. Importa *qué* clase se confunde con cuál.
- **Precision** (de lo que predije positivo, cuánto acerté) vs **Recall** (de los positivos reales, cuántos capturé): hay *trade-off*. **F1** los combina; **F1-macro** promedia por clase sin que la mayoritaria domine.
- **ROC-AUC** (binaria): probabilidad de rankear un positivo por encima de un negativo. 0.5 = azar, 1.0 = perfecto; robusta al desbalance.

**Qué dicen los números reales (test, 114 casos no vistos)**: el modelo elegido (regresión logística) alcanza **accuracy 0.97 y ROC-AUC 0.997** — prácticamente un ranking perfecto. Por clase: en **clase 0 (maligno)** precision 0.98, recall 0.95, F1 0.96 sobre 43 casos; en **clase 1 (benigno)** precision 0.97, recall 0.99, F1 0.98 sobre 71 casos. La matriz de confusión muestra que **el único error que pesa son ~2 malignos clasificados como benignos** (recall 0.95 sobre 43 reales = 2 falsos negativos), más algún benigno marcado como maligno.

**Qué se infiere**: clínicamente lo grave es el falso negativo (decirle "benigno" a un tumor maligno), y ahí el recall de la clase maligna es 0.95 — bueno pero no perfecto, el punto a vigilar si esto fuera a producción (bajar el umbral subiría ese recall a costa de más falsas alarmas). Como **el baseline ≈ la red** (0.995 vs 0.983 en CV), el problema es mayormente lineal y no pedía una red neuronal. **Limitación**: hiperparámetros sin tunear; con `GridSearchCV` y ajuste de umbral se podría empujar el recall maligno sin sacrificar casi nada de AUC.


## ✅ Conclusiones

- El EDA sobre **Breast Cancer** confirmó un dataset **limpio y de alta señal**: 569 muestras × 30 features numéricas, **0 faltantes**, target desbalanceado 62.7% benigno / 37.3% maligno (ratio 0.59). Las variables más predictivas son `worst concave points` (r = −0.79) y el bloque tamaño (`worst perimeter`, `mean concave points`), con fuerte multicolinealidad interna (`mean radius` ~ `mean perimeter` = 0.998).
