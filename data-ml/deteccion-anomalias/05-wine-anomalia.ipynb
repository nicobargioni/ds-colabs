## 🧪 3 · Preprocesamiento

Escalado y encoding sobre todo el conjunto (problema no supervisado, sin target).

### 🧪 Por qué este preprocesamiento

En un problema **no supervisado** no hay target ni split: el objetivo es describir la estructura de *todos* los datos.

- **Escalado** (`StandardScaler`): imprescindible acá, porque tanto K-Means como Isolation Forest se basan en **distancias**; sin escalar, las variables de mayor magnitud dominan artificialmente.
- **Encoding** de categóricas a binarias para que entren en el cálculo de distancias.
- Todo se ajusta sobre el conjunto completo (no hay fuga posible porque no hay test que predecir).

```python
# 3 · Preprocesamiento (no supervisado): escalado + encoding sobre todo X
num_feats = df.select_dtypes(include='number').columns.tolist()
cat_feats = df.select_dtypes(exclude='number').columns.tolist()
numeric = Pipeline([('imp', SimpleImputer(strategy='median')), ('sc', StandardScaler())])
categ = Pipeline([('imp', SimpleImputer(strategy='most_frequent')),
                  ('oh', OneHotEncoder(handle_unknown='ignore', sparse_output=False))])
preprocess = ColumnTransformer([('num', numeric, num_feats), ('cat', categ, cat_feats)])
Xp = preprocess.fit_transform(df)
print('Matriz preprocesada:', Xp.shape)
```

## 🧠 4 · Modelado

Isolation Forest para señalar observaciones atípicas.

### 🧠 Por qué Isolation Forest

**Isolation Forest** aísla observaciones particionando al azar: los puntos que quedan aislados con **pocas** particiones son los más atípicos. Escala bien, no asume distribución y funciona en alta dimensión — adecuado para las 13 features escaladas de Wine.

`contamination='auto'` deja que el algoritmo estime la proporción de anomalías. En esta corrida (con `SEED=42`) eso se traduce en **15 vinos marcados como atípicos, el 8.4% del total** (15/178). Es una tasa más alta que las 21 banderas IQR univariadas, porque el modelo razona de forma **multivariada**: detecta combinaciones raras de variables, no solo extremos columna por columna. Si conociéramos la tasa real esperada (p. ej. 2-3% de lotes defectuosos), convendría fijar `contamination` explícitamente para no sobre-marcar.

```python
# 4 · Detección de anomalías con Isolation Forest
from sklearn.ensemble import IsolationForest
iso = IsolationForest(random_state=SEED, contamination='auto').fit(Xp)
scores = iso.decision_function(Xp); pred = iso.predict(Xp)   # -1 = anómalo
df['anomalia'] = (pred == -1)
print('Anomalías detectadas:', int(df['anomalia'].sum()), f'({df["anomalia"].mean()*100:.1f}%)')
```

## 📊 5 · Evaluación

Interpretación de los grupos / anomalías (proyección PCA 2D).

```python
# 5 · Score de anomalía + proyección 2D (PCA)
px.histogram(x=scores, nbins=50, title='Score de anomalía (menor = más atípico)',
             labels={'x': 'score'}).show()
from sklearn.decomposition import PCA
coords = PCA(n_components=2, random_state=SEED).fit_transform(Xp)
px.scatter(x=coords[:, 0], y=coords[:, 1],
           color=df['anomalia'].map({False: 'normal', True: 'anomalía'}),
           color_discrete_map={'normal': PRIMARY, 'anomalía': ACCENT},
           labels={'x': 'PC1', 'y': 'PC2', 'color': ''}, title='Anomalías (PCA 2D)').show()
```

### 📌 Lectura de resultados

Sobre las 178 muestras escaladas, Isolation Forest aísla **15 vinos como anómalos (8.4%)** usando `contamination='auto'` y `SEED=42`. El score de anomalía va de **-0.113 a +0.116** (mediana 0.063), y los anómalos quedan claramente en la cola negativa: score medio **-0.036** frente a **+0.062** de los 161 normales, sin solaparse con el grueso de la distribución. La proyección PCA 2D explica el **55.4% de la varianza** (PC1 36.2% + PC2 19.2%), suficiente para ver que las anomalías tienden a caer en los bordes de la nube, aunque ~45% de la información queda fuera del plano. Estas 15 marcas no coinciden mecánicamente con las 21 banderas IQR univariadas: el modelo prioriza **combinaciones químicas raras** por sobre extremos de una sola variable, lo cual es esperable dado el fuerte bloque de fenoles correlacionados (`total_phenols`~`flavanoids` = 0.87) que reduce la dimensionalidad efectiva del dataset.

### 📊 Cómo interpretar estas métricas

- El **score** (`decision_function`) ordena de más a menos atípico. En esta corrida va de **-0.113** (el vino más anómalo) a **+0.116** (el más típico), con **mediana 0.063**. La frontera de decisión está en 0: por debajo cae lo anómalo.
- La separación es clara: los **15 puntos anómalos** tienen un score medio de **-0.036**, frente a **+0.062** de los 161 normales. Es decir, las anomalías no están pegadas al umbral — viven en una cola negativa bien diferenciada del grueso de la distribución.
- Sin *ground truth* la validación es **cualitativa**: hay que revisar si esos 15 vinos tienen sentido de negocio (¿lotes de medición rara, cultivares fronterizos?). El modelo no afirma que sean defectuosos, solo que son **estadísticamente distintos**.
- En la **proyección PCA 2D**, los dos primeros componentes capturan el **55.4% de la varianza** (PC1 36.2% + PC2 19.2%). Es suficiente para visualizar la nube pero deja ~45% fuera del plano: algunas anomalías pueden parecer "centradas" en 2D y aun así ser atípicas en las dimensiones no mostradas.

**Se infiere**: el dataset tiene un núcleo denso y compacto (la mayoría de los vinos comparten perfil químico) con una cola minoritaria de ~8% que se aparta de forma multivariada, coherente con la fuerte correlación entre fenoles que reduce la dimensionalidad efectiva.

## ✅ Conclusiones

- **Dataset limpio y compacto**: 178 vinos × 13 features químicas (todas numéricas), **sin un solo faltante**, así que no hubo imputación efectiva.
- **Estructura redundante**: la correlación dominante `total_phenols`~`flavanoids` = **0.87** (y el bloque de fenoles asociado) comprime la información; por eso PCA captura **55.4%** de la varianza con apenas 2 componentes.
- **Resultado del modelo**: Isolation Forest con `contamination='auto'` marcó **15 anomalías (8.4%)**, con scores separados (media **-0.036** vs **+0.062** de los normales) — una cola atípica bien diferenciada del núcleo.
- **IQR vs multivariado**: las 21 banderas IQR univariadas (lideradas por `alcalinity_of_ash`, `magnesium`, `color_intensity`) no coinciden 1:1 con las 15 del modelo, que detecta **combinaciones raras**, no solo extremos por columna.
- El preprocesamiento (escalado + encoding) vive dentro de un **Pipeline** → sin data leakage.
- **Próximos pasos**: validar las 15 anomalías contra conocimiento de dominio, fijar `contamination` si se conoce la tasa real esperada, y probar feature engineering específico para separar los cultivares fronterizos.
