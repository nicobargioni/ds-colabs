## 🛠 0 · Setup

Imports y configuracion (Gymnasium + NumPy + Plotly). Semilla fija para reproducibilidad.

```python
import warnings; warnings.filterwarnings('ignore')
import numpy as np, pandas as pd
import gymnasium as gym
import plotly.express as px, plotly.graph_objects as go, plotly.io as pio
pio.templates.default = 'plotly_dark'
SEED = 42
rng = np.random.default_rng(SEED)
PRIMARY, ACCENT = '#7fd4ff', '#ff9f5c'
print('Setup OK · gymnasium', gym.__version__, '· numpy', np.__version__)
print('Q-learning tabular sobre el entorno Taxi (Toy Text)')
```

## 🚕 1 · El entorno Taxi

Estados, acciones y como se codifica la situacion del taxi.

```python
# El entorno Taxi: 500 estados discretos, 6 acciones discretas.
# (En gymnasium 1.x el id es 'Taxi-v4'; v3 quedo deprecado pero la dinamica es la misma.)
env = gym.make('Taxi-v4')
n_states  = env.observation_space.n
n_actions = env.action_space.n
print(f'Estados (observation_space): {n_states}')
print(f'Acciones (action_space):     {n_actions}')

# Decodificamos un estado para ver su estructura interna.
# El estado codifica: (fila_taxi, col_taxi, ubicacion_pasajero, destino)
obs, info = env.reset(seed=SEED)
taxi_row, taxi_col, pass_loc, dest = env.unwrapped.decode(obs)
print(f'\nEstado inicial = {obs}  ->  taxi=({taxi_row},{taxi_col}), '
      f'pasajero={pass_loc}, destino={dest}')
print('5 filas x 5 cols x 5 ubic.pasajero (4 paradas + dentro del taxi) x 4 destinos = 500')

# Las 6 acciones
acciones = {0:'Sur (abajo)', 1:'Norte (arriba)', 2:'Este (derecha)',
            3:'Oeste (izquierda)', 4:'Recoger pasajero', 5:'Dejar pasajero'}
print('\nAcciones:', acciones)
```

## 🎯 2 · Recompensas y baseline

El esquema de recompensas que guia el aprendizaje, y cuanto rinde no aprender (agente aleatorio).

```python
# Esquema de recompensas del entorno (clave para entender que va a aprender el agente):
#   -1   por cada paso (penaliza demorar -> empuja al camino corto)
#  +20   por dejar al pasajero en el destino correcto (objetivo)
#  -10   por recoger/dejar de forma ilegal (penaliza la chapuza)
print('Recompensas:')
print('  -1   por paso             -> incentiva resolver rapido')
print('  +20  dropoff correcto     -> el objetivo del episodio')
print('  -10  pickup/dropoff ilegal-> castiga acciones invalidas')

# Cuanto rinde NO aprender: un agente aleatorio sobre 1000 episodios (baseline).
def correr_aleatorio(env, episodios=1000, max_pasos=200, seed=SEED):
    g = np.random.default_rng(seed)
    recs, pasos = [], []
    for ep in range(episodios):
        obs, _ = env.reset(seed=int(seed + ep))
        total, n = 0.0, 0
        for _ in range(max_pasos):
            a = int(g.integers(n_actions))
            obs, r, term, trunc, _ = env.step(a)
            total += r; n += 1
            if term or trunc:
                break
        recs.append(total); pasos.append(n)
    return np.array(recs), np.array(pasos)

rec_rand, pasos_rand = correr_aleatorio(env)
print(f'\nBASELINE (agente aleatorio, 1000 episodios):')
print(f'  recompensa media = {rec_rand.mean():.2f}')
print(f'  pasos medios     = {pasos_rand.mean():.1f}')
```

**Por que Taxi es un buen banco de pruebas para RL tabular.** El espacio de estados es
**finito y chico** (500), asi que toda la funcion de valor cabe en una tabla `Q[estado, accion]`
de 500x6 = 3000 numeros. No hace falta una red neuronal: podemos guardar *un valor por cada
combinacion estado-accion* y actualizarlo directamente. Eso lo vuelve el ejemplo canonico para
entender la mecanica de Q-learning antes de saltar a Deep RL.

La recompensa esta diseñada para que **el comportamiento optimo emerja solo**: el `-1` por paso
empuja al camino corto, el `+20` premia cumplir el objetivo y el `-10` castiga las acciones
ilegales (intentar dejar al pasajero donde no corresponde). El agente no recibe ninguna
instruccion de *como* manejar; solo siente estas recompensas y tiene que descubrir la politica.

## 🧠 3 · Q-learning

Tabla Q de 500x6, entrenamiento por episodios con epsilon-greedy y decay de epsilon.

```python
# ----------------------------------------------------------------------
# Q-LEARNING TABULAR
# Regla de actualizacion (Bellman, TD off-policy):
#   Q(s,a) <- Q(s,a) + alpha * [ r + gamma * max_a' Q(s',a')  -  Q(s,a) ]
#                              \_______________ TD target ______________/
# ----------------------------------------------------------------------
ALPHA      = 0.10     # tasa de aprendizaje: cuanto pesa cada nueva experiencia
GAMMA      = 0.99     # factor de descuento: cuanto valen las recompensas futuras
EPS_INI    = 1.0      # epsilon inicial: 100% exploracion al arrancar
EPS_FIN    = 0.01     # epsilon final: casi pura explotacion al final
EPS_DECAY  = 0.9995   # decay multiplicativo por episodio
EPISODIOS  = 20000
MAX_PASOS  = 200

Q = np.zeros((n_states, n_actions))     # tabla Q inicializada en cero
g = np.random.default_rng(SEED)         # RNG para la exploracion epsilon-greedy
eps = EPS_INI
recompensas_ep, pasos_ep, eps_ep = [], [], []

for ep in range(EPISODIOS):
    obs, _ = env.reset(seed=int(SEED + ep))
    total, n = 0.0, 0
    for _ in range(MAX_PASOS):
        # epsilon-greedy: con prob. eps exploro al azar, si no exploto el mejor Q
        if g.random() < eps:
            a = int(g.integers(n_actions))
        else:
            a = int(np.argmax(Q[obs]))
        nobs, r, term, trunc, _ = env.step(a)
        # actualizacion de Bellman
        td_target = r + GAMMA * np.max(Q[nobs]) * (not term)
        Q[obs, a] += ALPHA * (td_target - Q[obs, a])
        obs = nobs; total += r; n += 1
        if term or trunc:
            break
    recompensas_ep.append(total); pasos_ep.append(n); eps_ep.append(eps)
    eps = max(EPS_FIN, eps * EPS_DECAY)   # decay de epsilon

recompensas_ep = np.array(recompensas_ep)
pasos_ep = np.array(pasos_ep)
print(f'Entrenamiento terminado: {EPISODIOS} episodios')
print(f'epsilon final = {eps:.4f}  (arranco en {EPS_INI})')
print(f'recompensa media primeros 100 ep: {recompensas_ep[:100].mean():.1f}')
print(f'recompensa media ultimos  100 ep: {recompensas_ep[-100:].mean():.1f}')
```

**Por que Q-learning.** Es el algoritmo *off-policy* mas directo para aprender control sin
modelo del entorno: no necesitamos conocer de antemano las probabilidades de transicion ni las
recompensas, solo *muestrearlas* actuando. La tabla `Q[s,a]` estima **cuanta recompensa futura
total esperamos** si en el estado `s` tomamos la accion `a` y despues jugamos optimo. Una vez
aprendida, la politica es trivial: en cada estado, elegir `argmax Q[s, ·]`.

**Que hace cada hiperparametro:**

- **alpha (0.10) — tasa de aprendizaje.** Cuanto corregimos `Q(s,a)` hacia la nueva evidencia en
  cada paso. Alta = aprende rapido pero inestable (ruido); baja = estable pero lento. 0.10 es un
  punto medio robusto para entornos deterministas como Taxi.
- **gamma (0.99) — factor de descuento.** Cuanto valen las recompensas futuras frente a la
  inmediata. Cerca de 1 (0.99) hace al agente **paciente**: prioriza el `+20` del final aunque
  cueste varios `-1` por el camino. Si fuera bajo, el agente seria miope y no "veria" la
  recompensa lejana del dropoff.
- **epsilon — tasa de exploracion.** Probabilidad de tomar una accion *al azar* en vez de la
  mejor conocida. Resuelve el dilema **explorar vs explotar**: si siempre explotara, se quedaria
  atrapado en lo primero que parezca bueno; si siempre explorara, nunca aprovecharia lo aprendido.

**El decay de epsilon es el mecanismo central.** Arrancamos en `eps=1.0` (100% exploracion: el
agente no sabe nada del mapa, conviene probar todo) y lo bajamos multiplicando por `0.9995` cada
episodio hasta un piso de `0.01`. La intuicion: **al principio explorar paga** (hay que conocer
el entorno), **al final explotar paga** (ya sabemos el camino, toca cobrarlo). El TD target usa
`max_a' Q(s',a')` —el *mejor* valor del siguiente estado, no la accion que efectivamente
tomamos— y por eso Q-learning es *off-policy*: aprende la politica optima incluso mientras se
comporta de forma exploratoria.

## 📈 4 · Curva de aprendizaje

Recompensa por episodio (media movil): la firma visual de que el agente esta aprendiendo.

```python
# Curva de aprendizaje: recompensa por episodio con media movil (ventana 100).
# La media movil suaviza la varianza enorme episodio-a-episodio y deja ver la tendencia.
def media_movil(x, w=100):
    return pd.Series(x).rolling(w, min_periods=1).mean().values

mm_rec = media_movil(recompensas_ep, 100)
x = np.arange(EPISODIOS)

fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=recompensas_ep, mode='lines', name='recompensa cruda',
                         line=dict(color='rgba(127,212,255,0.18)', width=1)))
fig.add_trace(go.Scatter(x=x, y=mm_rec, mode='lines', name='media movil (100 ep)',
                         line=dict(color=PRIMARY, width=2.5)))
fig.add_hline(y=rec_rand.mean(), line_dash='dash', line_color=ACCENT,
              annotation_text=f'baseline aleatorio ({rec_rand.mean():.0f})',
              annotation_position='bottom right')
fig.update_layout(title='Curva de aprendizaje · recompensa por episodio (Q-learning)',
                  xaxis_title='Episodio', yaxis_title='Recompensa total del episodio',
                  legend=dict(orientation='h', y=-0.2))
fig.show()

# ¿En que episodio "converge"? Primer episodio cuya media movil supera +5 (umbral de exito)
umbral = 5.0
sup = np.where(mm_rec >= umbral)[0]
ep_conv = int(sup[0]) if len(sup) else None
print(f'Recompensa cruda: min={recompensas_ep.min():.0f}, max={recompensas_ep.max():.0f}')
print(f'Media movil final (ultimos 100 ep): {mm_rec[-1]:.2f}')
print(f'Converge (media movil cruza +{umbral:.0f}) en el episodio: {ep_conv}')
```

**Como leer la curva.** El trazo tenue es la recompensa cruda episodio-a-episodio: tiene una
varianza enorme porque cada episodio arranca en un estado aleatorio y, sobre todo al principio,
la exploracion mete acciones al azar. El trazo grueso es la **media movil de 100 episodios**, que
filtra ese ruido y deja ver la **tendencia real de aprendizaje**: la curva sube desde valores
fuertemente negativos (el agente choca, hace pickups ilegales, agota los pasos) hacia un plateau
positivo, donde ya resuelve casi siempre por el camino corto. La linea punteada marca el baseline
aleatorio: la distancia entre el plateau y esa linea es, literalmente, *cuanto vale haber
aprendido*.

## 🏁 5 · Evaluacion

Politica greedy sobre 1000 episodios vs baseline aleatorio: recompensa media y pasos medios.

```python
# Evaluacion: politica GREEDY (epsilon=0, sin exploracion) sobre 1000 episodios.
def evaluar_greedy(env, Q, episodios=1000, max_pasos=200, seed=SEED):
    recs, pasos, exitos = [], [], 0
    for ep in range(episodios):
        obs, _ = env.reset(seed=int(seed + 10000 + ep))
        total, n = 0.0, 0
        for _ in range(max_pasos):
            a = int(np.argmax(Q[obs]))          # pura explotacion
            obs, r, term, trunc, _ = env.step(a)
            total += r; n += 1
            if term or trunc:
                if r == 20:                      # dropoff correcto
                    exitos += 1
                break
        recs.append(total); pasos.append(n)
    return np.array(recs), np.array(pasos), exitos

rec_greedy, pasos_greedy, exitos = evaluar_greedy(env, Q)

comp = pd.DataFrame({
    'Agente': ['Aleatorio (baseline)', 'Q-learning (greedy)'],
    'Recompensa media': [rec_rand.mean(), rec_greedy.mean()],
    'Pasos medios': [pasos_rand.mean(), pasos_greedy.mean()],
}).round(2)
print('COMPARACION sobre 1000 episodios de evaluacion:\n')
print(comp.to_string(index=False))
print(f'\nEpisodios resueltos con exito (dropoff correcto): {exitos}/1000 '
      f'= {100*exitos/1000:.1f}%')
print(f'Mejora de recompensa vs baseline: '
      f'{rec_greedy.mean() - rec_rand.mean():+.1f} puntos')

fig = px.bar(comp, x='Agente', y='Recompensa media', color='Agente', text='Recompensa media',
             color_discrete_map={'Aleatorio (baseline)': ACCENT, 'Q-learning (greedy)': PRIMARY},
             title='Recompensa media: Q-learning vs agente aleatorio (1000 episodios)')
fig.update_traces(textposition='outside'); fig.update_layout(showlegend=False)
fig.show()
```

**Por que evaluar con epsilon=0.** Durante el entrenamiento el agente explora (toma acciones
al azar de vez en cuando), lo cual ensucia la recompensa. Para medir *que tan buena es la politica
aprendida* apagamos la exploracion: en cada estado tomamos siempre `argmax Q[s, ·]`. Esa es la
**politica greedy**, la que de verdad usariamos en produccion. La comparamos contra el agente
aleatorio sobre el *mismo numero de episodios* (1000) para que el contraste sea justo: misma vara,
dos politicas.

|---|---|---|---|
| Aleatorio (baseline) | **-776.33** | 197.3 | ~0% |
| Q-learning (greedy) | **+7.89** | **13.1** | **1000/1000 (100%)** |

**Que dicen los numeros:**

- **La recompensa salta de -776.33 a +7.89: una mejora de +784 puntos.** El baseline aleatorio
  practicamente nunca resuelve el episodio; acumula cientos de penalizaciones (`-1` por paso
  durante los 197 pasos hasta agotar el limite, mas `-10` por cada pickup/dropoff ilegal). El
  agente entrenado, en cambio, va directo: **resuelve los 1000 episodios** (100% de dropoffs
  correctos) en apenas **13 pasos de media**.

- **Por que +7.89 y no mas:** el optimo teorico de Taxi ronda recompensa media ~8, porque cada
  episodio paga inevitablemente unos cuantos `-1` por los pasos minimos necesarios antes de
  cobrar el `+20`. Estar en ~7.9 con dropoff correcto en el 100% de los casos indica que el
  agente aprendio **casi la politica optima**: no malgasta pasos ni comete acciones ilegales.

- **Los pasos caen de 197.3 a 13.1** (15x menos): la metrica mas tangible de que el agente
  encontro el camino corto. El `-1` por paso hizo su trabajo: penalizar la demora empujo a la
  politica hacia trayectorias eficientes.

**La curva confirma el aprendizaje.** La media movil arranca cerca de **-765** (primeros 100
episodios, casi todo exploracion al azar) y termina en **+7.39** (ultimos 100 episodios). El
cruce del umbral de exito (+5) ocurre alrededor del **episodio 5.461**: ese es el punto donde el
agente deja de penar y empieza a resolver con consistencia. El resto del entrenamiento afina la
politica y, con el epsilon ya bajo, la explota.

**Conexion con la hipotesis.** Se valida por completo: Q-learning con epsilon-greedy y decay
**convergio a una politica casi optima** que supera al agente aleatorio por +784 puntos, pasando
de recompensa fuertemente negativa a positiva estable. Y lo logro **sin ningun modelo previo del
entorno**: solo prueba, error y la regla de actualizacion de Bellman.

## ✅ Conclusiones

- Q-learning **aprende sin modelo del entorno**: solo prueba, error y la regla de actualizacion de Bellman bastan para hallar una politica casi optima.
- El **decay de epsilon** es el corazon del balance explorar/explotar: mucha exploracion al principio (conocer el mapa), casi pura explotacion al final (cobrar lo aprendido).
- La **curva de aprendizaje** muestra la transicion de recompensa fuertemente negativa a positiva estable: es la firma visual de que el agente dejo de penar.
